# üìò **Ejercicio 01**  

---

> ‚ú® **Objetivo:** Calcular y analizar la **productividad multifactorial** y la **productividad laboral** en un curso t√≠pico de **3 cr√©ditos** en la Universidad *Boehring*.  

---

## üîç **Datos del problema**  

üéì La matr√≠cula estudiantil en la Universidad *Boehring* es de **$150** por hora cr√©dito por semestre.  
üèõÔ∏è El estado subsidia la universidad con **$100** por hora cr√©dito por semestre.  
üìö El tama√±o promedio de un grupo en un curso de **3 cr√©ditos** es de **50 estudiantes**.  

### üí∞ **Costos asociados**  

- üßë‚Äçüè´ **Costos laborales**: **$4,000** por clase  
- üìñ **Costos de materiales**: **$20** por estudiante por clase  
- üè¢ **Costos generales** (_overhead_): **$25,000** por clase  

---

## ‚ùì **Problemas a resolver**  

### üèÜ **Parte A: Productividad multifactorial**  
üîπ ¬øCu√°l es la raz√≥n de productividad **multifactorial** para este proceso de curso?  

### üë©‚Äçüè´ **Parte B: Productividad laboral**  
üìÜ Si los instructores trabajan un promedio de **14 horas por semana** durante **16 semanas** para cada clase de **3 cr√©ditos** con **50 estudiantes**, ¬øcu√°l es la raz√≥n de productividad **laboral**?  

---

## üìä **Tabla 1: Datos del problema**  

| üî¢ **Variable**                     | üìñ **Descripci√≥n**                                      | üí≤ **Valor**   |
|--------------------------------------|---------------------------------------------------------|--------------:|
| üéì **Cuota escolar por hora cr√©dito**  | Costo que pagan los estudiantes por hora cr√©dito       | **$150**      |
| üèõÔ∏è **Apoyo estatal por hora cr√©dito**  | Cantidad que aporta el gobierno estatal por hora cr√©dito | **$100**      |
| üë®‚Äçüéì **Tama√±o promedio del grupo**       | N√∫mero de estudiantes por grupo                        | **50**        |
| üìö **N√∫mero de cr√©ditos por curso**   | N√∫mero de cr√©ditos por curso                           | **3**         |

> üìù **Nota:** Se debe calcular la productividad teniendo en cuenta todos los costos y beneficios asociados al curso.  

---


```r
# Definici√≥n de variables
cuota_escolar <- 150  # USD por hora cr√©dito
apoyo_estatal <- 100   # USD por hora cr√©dito
tamano_grupo <- 50     # N√∫mero de estudiantes
creditos_curso <- 3    # N√∫mero de cr√©ditos por curso

costo_mano_obra <- 4000    # USD por clase
costo_materiales <- 20     # USD por estudiante
costos_generales <- 25000  # USD por clase
horas_trabajadas <- 14 * 16  # 14 horas/semana por 16 semanas

# C√°lculo del valor del producto (OUTPUTS)
cuota_escolar_total <- cuota_escolar * creditos_curso * tamano_grupo
apoyo_estatal_total <- apoyo_estatal * creditos_curso * tamano_grupo
valor_producto_total <- cuota_escolar_total + apoyo_estatal_total

# C√°lculo del valor de los insumos (INPUTS)
costo_materiales_total <- costo_materiales * tamano_grupo
valor_insumos_total <- costo_mano_obra + costo_materiales_total + costos_generales

# C√°lculo de la productividad multifactorial
productividad_multifactorial <- valor_producto_total / valor_insumos_total

# C√°lculo de la productividad laboral
productividad_laboral <- valor_producto_total / horas_trabajadas

# Impresi√≥n de resultados
cat("=== Resultados del Ejercicio ===\n\n")

cat("Valor del Producto (Outputs):\n")
cat("Cuota Escolar Total: $", cuota_escolar_total, "\n")
cat("Apoyo Estatal Total: $", apoyo_estatal_total, "\n")
cat("Valor Total del Producto: $", valor_producto_total, "\n\n")

cat("Valor de los Insumos (Inputs):\n")
cat("Costo de Mano de Obra: $", costo_mano_obra, "\n")
cat("Costo de Materiales Total: $", costo_materiales_total, "\n")
cat("Costos Generales: $", costos_generales, "\n")
cat("Valor Total de los Insumos: $", valor_insumos_total, "\n\n")

cat("Productividad Multifactorial:\n")
cat("Productividad = Valor del Producto / Valor de los Insumos = ", productividad_multifactorial, "\n\n")

cat("Productividad Laboral:\n")
cat("Horas trabajadas por maestro: ", horas_trabajadas, " horas\n")
cat("Ingresos por curso: $", valor_producto_total, "\n")
cat("Productividad Laboral (USD por hora): ", productividad_laboral, "\n")
```

# üëó **Ejercicio 02: Productividad Laboral en la Industria Textil**  

---

> üìå **Objetivo:** Calcular y analizar la **productividad laboral** en la producci√≥n de prendas de vestir en *Natalie Attire*.  

---

## üè≠ **Descripci√≥n del proceso de manufactura**  

üëï *Natalie Attire* es una empresa de moda que fabrica prendas de vestir.  
üìÖ Durante una semana en particular:  

- ‚è≥ **360 horas** de trabajo fueron empleadas.  
- üßµ Se produjo un lote de **132 prendas**, de las cuales:  
  - ‚ùå **52 prendas** eran de **segunda** (con defectos).  
  - ‚úÖ **80 prendas** fueron vendidas a precio completo.  

### üí∞ **Precios de venta**  

- üè∑Ô∏è **Prendas de segunda**: **$90** cada una (vendidas en la tienda de f√°brica).  
- üõçÔ∏è **Prendas sin defectos**: **$200** cada una (vendidas en distribuci√≥n minorista).  

---

## ‚ùì **Problema a resolver**  

### üìä **C√°lculo de productividad laboral**  
üîπ ¬øCu√°l es la **raz√≥n de productividad laboral** (_labor productivity ratio_) en este proceso de manufactura?  

---

## üìä **Tabla 1: Datos del problema**  

| üî¢ **Variable**                        | üí≤ **Valor**   |
|----------------------------------------|--------------:|
| üëï **N√∫mero total de prendas**          | **132**       |
| ‚ùå **N√∫mero de prendas de segunda**     | **52**        |
| ‚úÖ **N√∫mero de prendas restantes**      | **80**        |
| üè∑Ô∏è **Precio de venta por prenda de segunda** | **$90**        |
| üõçÔ∏è **Precio de venta por prenda restante**  | **$200**       |
| ‚è≥ **Horas de trabajo empleadas**       | **360**       |

> üìù **Nota:** Para calcular la productividad laboral, se debe considerar el valor total de producci√≥n generado en funci√≥n de las horas de trabajo utilizadas.  

---


```r
# Definir variables
num_prendas_totales <- 132
num_prendas_segunda <- 52
num_prendas_buenas <- 80
precio_segunda <- 90
precio_buenas <- 200
horas_trabajo <- 360

# C√°lculo del valor total del producto
valor_prendas_segunda <- num_prendas_segunda * precio_segunda
valor_prendas_buenas <- num_prendas_buenas * precio_buenas
valor_total_producto <- valor_prendas_segunda + valor_prendas_buenas

# C√°lculo de la productividad laboral
productividad_laboral <- valor_total_producto / horas_trabajo

# Mostrar resultados
cat("---- Resultados ----\n")
cat("Valor total de prendas de segunda: $", valor_prendas_segunda, "\n")
cat("Valor total de prendas buenas: $", valor_prendas_buenas, "\n")
cat("Valor total del producto: $", valor_total_producto, "\n")
cat("Horas de trabajo empleadas:", horas_trabajo, "horas\n")
cat("Productividad laboral: $", round(productividad_laboral, 2), "por hora\n")
```

üîª **Problemas actuales:**

- üìà Para mejorar el reclutamiento, los **costos generales** (_overhead_) aumentar√°n de **$25,000** a **$30,000** por secci√≥n de clase.

- üí∞ **Costos laborales** aumentar√°n a **$6,500** por curso de 3 cr√©ditos.  
- üìö **Costos de materiales** ser√°n de **$25** por estudiante por curso.  
- üéì **Matr√≠cula** aumentar√° a **$200** por cr√©dito, con un **apoyo estatal** de **$100** por cr√©dito.  

---

## ‚ùì **Preguntas a Resolver**  

### üî¢ **Parte A: Productividad Multifactorial**  
üìä ¬øCu√°l es el **√≠ndice de productividad multifactorial** despu√©s de estos cambios?  
üìâ En comparaci√≥n con el resultado del **Problema Resuelto 1**, ¬øla productividad **aument√≥ o disminuy√≥**?  

### üë®‚Äçüè´ **Parte B: Productividad Laboral**  
‚è≥ Si los instructores trabajan **20 horas por semana** durante **16 semanas** para cada curso de **3 cr√©ditos** con **75 estudiantes**,  
¬øCu√°l ser√° el **√≠ndice de productividad laboral**?  

---

## üìä **Tabla 1: Datos del Problema**  

| üî¢ **Variable**                             | üí≤ **Valor**  |
|---------------------------------------------|-------------:|
| üèõÔ∏è **Costo general por secci√≥n (nuevo)**    | **$30,000**  |
| üèõÔ∏è **Costo general por secci√≥n (anterior)** | **$25,000**  |
| üéì **Tama√±o promedio de la clase**          | **75**       |
| üíº **Costos laborales por curso**           | **$6,500**   |
| üìñ **Costos de materiales por estudiante**  | **$25**      |
| üéüÔ∏è **Costo de matr√≠cula por cr√©dito**       | **$200**     |
| üèõÔ∏è **Apoyo estatal por cr√©dito**            | **$100**     |
| üìö **Cr√©ditos por curso**                   | **3**        |
| ‚è≥ **Horas de trabajo del instructor**      | **320**      |

> üìù **Nota:** Se analizar√° el impacto de estos cambios en la eficiencia del proceso educativo y en el uso de recursos.  

---



```r
# ========================================================
# C√°lculo de Productividad Laboral y Multifactorial en R
# ========================================================
# Administraci√≥n de Operaciones: La productividad mide la eficiencia 
# en el uso de recursos para generar productos o servicios.
# En este ejercicio se analizan cambios en los costos y horas trabajadas,
# evaluando el impacto en la productividad.

# Definici√≥n de variables iniciales
# ----------------------------------

# Costos y par√°metros originales
cuota_escolar_original <- 150
apoyo_estatal_original <- 100
tamano_grupo_original <- 50
creditos_curso <- 3
costo_mano_obra_original <- 4000
costo_generales_original <- 25000
costo_materiales_original <- 1000
horas_trabajo_original <- 224

# Costos y par√°metros ajustados
cuota_escolar_ajuste <- 200
apoyo_estatal_ajuste <- 100
tamano_grupo_ajuste <- 75
costo_mano_obra_ajuste <- 6500
costo_generales_ajuste <- 30000
costo_materiales_ajuste <- 1875
horas_trabajo_ajuste <- 320

# C√°lculo del valor del producto (Ingresos generados)
# ----------------------------------------------------
valor_producto_original <- (cuota_escolar_original + apoyo_estatal_original) * creditos_curso * tamano_grupo_original
valor_producto_ajuste <- (cuota_escolar_ajuste + apoyo_estatal_ajuste) * creditos_curso * tamano_grupo_ajuste

# C√°lculo de los insumos (Costos totales)
# ---------------------------------------
insumos_original <- costo_mano_obra_original + costo_generales_original + costo_materiales_original
insumos_ajuste <- costo_mano_obra_ajuste + costo_generales_ajuste + costo_materiales_ajuste

# C√°lculo del √≠ndice de productividad multifactorial
# --------------------------------------------------
productividad_multifactorial_original <- valor_producto_original / insumos_original
productividad_multifactorial_ajuste <- valor_producto_ajuste / insumos_ajuste

# C√°lculo de la productividad laboral (por hora trabajada)
# --------------------------------------------------------
productividad_laboral_original <- valor_producto_original / horas_trabajo_original
productividad_laboral_ajuste <- valor_producto_ajuste / horas_trabajo_ajuste

# C√°lculo del cambio porcentual en la productividad
# -------------------------------------------------
cambio_productividad_multifactorial <- ((productividad_multifactorial_ajuste - productividad_multifactorial_original) / productividad_multifactorial_original) * 100
cambio_productividad_laboral <- ((productividad_laboral_ajuste - productividad_laboral_original) / productividad_laboral_original) * 100

# Resultados
# ==========
cat("\n--- Resultados de Productividad ---\n")
cat("Valor del Producto (Original):", valor_producto_original, "\n")
cat("Valor del Producto (Ajustado):", valor_producto_ajuste, "\n")
cat("Productividad Multifactorial Original:", round(productividad_multifactorial_original, 4), "\n")
cat("Productividad Multifactorial Ajustada:", round(productividad_multifactorial_ajuste, 4), "\n")
cat("Cambio en Productividad Multifactorial:", round(cambio_productividad_multifactorial, 2), "%\n")
cat("Productividad Laboral Original:", round(productividad_laboral_original, 4), "\n")
cat("Productividad Laboral Ajustada:", round(productividad_laboral_ajuste, 4), "\n")
cat("Cambio en Productividad Laboral:", round(cambio_productividad_laboral, 2), "%\n")

# ========================================================
# CONCLUSI√ìN:
# - Se observa un incremento en la productividad laboral y multifactorial.
# - El aumento de costos fue compensado por un mayor tama√±o de grupo y eficiencia.
# - Es crucial evaluar continuamente estos indicadores para optimizar recursos.
# ========================================================
```

# üß∫ **Ejercicio 04: Productividad Laboral en Suds and Duds Laundry**  

---

> üìå **Objetivo:** Analizar la **productividad laboral** en el servicio de lavander√≠a *Suds and Duds Laundry*.  

---

## üè≠ **Descripci√≥n del Proceso**  

üëî *Suds and Duds Laundry* es un servicio de lavander√≠a que se especializa en el **lavado y planchado de camisas de vestir**.  
üìÖ Se ha registrado la producci√≥n semanal durante **cinco semanas**, considerando el **n√∫mero de trabajadores, total de horas trabajadas y camisas procesadas**.  

---

## üìä **Datos de Producci√≥n**  

| üìÖ **Semana** | üë• **Trabajadores**  | ‚è≥ **Total de horas** | üëî **Camisas procesadas** |
|-------------|--------------------|---------------------:|-------------------------:|
| 1           | **Sud y Dud**       | **24**              | **68**                   |
| 2           | **Sud y Dud**       | **46**              | **130**                  |
| 3           | **Sud, Dud y Jud**  | **62**              | **152**                  |
| 4           | **Sud, Dud y Jud**  | **51**              | **125**                  |
| 5           | **Dud y Jud**       | **45**              | **131**                  |

---

## ‚ùì **Preguntas a Resolver**  

### üî¢ **Parte A: C√°lculo de Productividad Laboral**  
üìä ¬øCu√°l es la **raz√≥n de productividad laboral** (_labor productivity ratio_) para cada semana?  

### üìà **Parte B: An√°lisis del Patr√≥n de Productividad**  
üîç ¬øQu√© **patr√≥n de productividad laboral** se puede identificar a partir de los datos?  
üìâ ¬øC√≥mo influye el n√∫mero de trabajadores y las horas trabajadas en la eficiencia del servicio?  

> üìù **Nota:** La productividad laboral se calcular√° como el n√∫mero de camisas procesadas por hora trabajada.  

---



```r
# -----------------------------------------------------------
# C√°lculo de la Productividad Laboral - Suds and Duds Laundry
# -----------------------------------------------------------
# La productividad laboral se calcula como:
# Productividad = Producci√≥n (output) / Horas de trabajo (input)

# Definir los datos del problema
semanas <- c(1, 2, 3, 4, 5)  # N√∫mero de semanas
trabajadores <- c(2, 2, 3, 3, 2)  # N√∫mero de trabajadores por semana
horas_trabajo <- c(24, 46, 62, 51, 45)  # Total de horas trabajadas por semana
produccion <- c(68, 130, 152, 125, 131)  # Camisas producidas por semana

# Calcular la productividad laboral (camisas por hora)
productividad <- produccion / horas_trabajo  

# Imprimir los resultados de productividad laboral
cat("Productividad Laboral (camisas por hora):\n")
for (i in 1:length(semanas)) {
  cat(sprintf("Semana %d: %.4f camisas/hora\n", semanas[i], productividad[i]))
}

# Analizar los cambios en productividad entre semanas
cambio_productividad <- c(NA, diff(productividad) / productividad[-length(productividad)] * 100)

# Imprimir los cambios porcentuales de productividad
cat("\nCambio porcentual en productividad entre semanas:\n")
for (i in 2:length(semanas)) {
  cat(sprintf("Semana %d vs Semana %d: %.2f%%\n", semanas[i], semanas[i-1], cambio_productividad[i]))
}

# An√°lisis de productividad laboral
cat("\nAn√°lisis de la productividad laboral:\n")
cat("La productividad laboral disminuye en la Semana 3 cuando hay 3 trabajadores.\n")
cat("Esto sugiere que el tercer trabajador no contribuye significativamente al rendimiento.\n")
cat("Podr√≠a deberse a restricciones de espacio, equipos o flujo de trabajo ineficiente.\n")
```

> üìå **Objetivo:** Determinar los **porcentajes de reducci√≥n** necesarios en los costos de **materiales, mano de obra y costos generales**  
para lograr una **mejora del 10% en la productividad multifactorial** en *White Tiger Electronics*.

## üè≠ **Descripci√≥n del Proceso**  

üìÄ *White Tiger Electronics* fabrica **reproductores de CD** utilizando un **proceso de ensamblaje automatizado**.

| üõ†Ô∏è **Materiales**   | **70**         |
| üè≠ **Costos generales** | **50**         |
| üí∞ **Total**         | **150**        |

üè∑Ô∏è **Precio de venta por unidad:** **$300**  

---

## ‚ùì **Preguntas a Resolver**  

### üî¢ **Parte A: Reducci√≥n de Costos en Materiales**  
üìâ ¬øEn qu√© **porcentaje deben reducirse los costos de materiales** para lograr una **mejora del 10% en la productividad multifactorial**?  

### üë∑ **Parte B: Reducci√≥n de Costos en Mano de Obra**  
üìâ ¬øEn qu√© **porcentaje deben reducirse los costos de mano de obra** para alcanzar la misma mejora en productividad?  

### üè≠ **Parte C: Reducci√≥n de Costos Generales**  
üìâ ¬øEn qu√© **porcentaje deben reducirse los costos generales** para alcanzar el objetivo de mejora?  

> üìù **Nota:** Se analizar√° la **relaci√≥n entre los costos totales y la productividad** para determinar el impacto de cada tipo de reducci√≥n.  

---


```r
# =========================================================
# C√°lculo de reducci√≥n de costos para mejorar productividad multifactorial
# Administraci√≥n de operaciones - An√°lisis de eficiencia en costos
# =========================================================

# Definici√≥n de variables
precio_venta <- 300       # Precio de venta por unidad ($)
costo_total <- 150        # Costo total de producci√≥n por unidad ($)
costo_mano_obra <- 30     # Costo de mano de obra por unidad ($)
costo_materiales <- 70    # Costo de materiales por unidad ($)
costo_overhead <- 50      # Costo de costos generales por unidad ($)
mejora_productividad <- 1.1  # Mejora del 10% en la productividad multifactorial

# C√°lculo de la nueva entrada de recursos requerida
nuevo_input <- costo_total / mejora_productividad

# C√°lculo de la reducci√≥n necesaria en costos
reduccion_necesaria <- costo_total - nuevo_input

# C√°lculo de reducci√≥n en Materiales
reduccion_materiales <- (reduccion_necesaria / costo_materiales) * 100
nuevo_costo_materiales <- costo_materiales - reduccion_necesaria

# C√°lculo de reducci√≥n en Mano de Obra
reduccion_mano_obra <- (reduccion_necesaria / costo_mano_obra) * 100
nuevo_costo_mano_obra <- costo_mano_obra - reduccion_necesaria

# C√°lculo de reducci√≥n en Overhead
reduccion_overhead <- (reduccion_necesaria / costo_overhead) * 100
nuevo_costo_overhead <- costo_overhead - reduccion_necesaria

# Impresi√≥n de resultados
cat("=======================================\n")
cat(" White Tiger Electronics - Resultados \n")
cat("=======================================\n\n")

cat("Costo total original de producci√≥n por unidad:", costo_total, "\n")
cat("Nuevo costo total necesario para lograr mejora del 10%:", round(nuevo_input, 4), "\n")
cat("Reducci√≥n total requerida en costos:", round(reduccion_necesaria, 4), "\n\n")

cat("Reducci√≥n necesaria en Materiales:\n")
cat(" - Nuevo costo de materiales:", round(nuevo_costo_materiales, 4), "\n")
cat(" - Porcentaje de reducci√≥n:", round(reduccion_materiales, 2), "%\n\n")

cat("Reducci√≥n necesaria en Mano de Obra:\n")
cat(" - Nuevo costo de mano de obra:", round(nuevo_costo_mano_obra, 4), "\n")
cat(" - Porcentaje de reducci√≥n:", round(reduccion_mano_obra, 2), "%\n\n")

cat("Reducci√≥n necesaria en Costos Generales (Overhead):\n")
cat(" - Nuevo costo de overhead:", round(nuevo_costo_overhead, 4), "\n")
cat(" - Porcentaje de reducci√≥n:", round(reduccion_overhead, 2), "%\n")
```

# üè≠ **Ejercicio 06: An√°lisis de Mejoras en el Proceso de Symtecks**  

---

> üìå **Objetivo:** Evaluar el impacto de las mejoras recientes en un proceso de **Symtecks**,  
utilizando **productividad multifactorial** y **productividad laboral** como indicadores de desempe√±o.  

---

## üè¢ **Descripci√≥n del Proceso**  

‚öôÔ∏è En **Symtecks**, un proceso espec√≠fico genera un **output** con un valor de **$100 por unidad producida**.  
üë∑‚Äç‚ôÇÔ∏è El **costo de la mano de obra** es de **$50 por hora**, incluyendo beneficios.  

üìä **El departamento de contabilidad** ha proporcionado informaci√≥n sobre la producci√≥n y costos en las √∫ltimas **cuatro semanas**.  

---

## üìä **Datos del Proceso**  

| üìÖ **Concepto**       | üîπ **Week 1** | üîπ **Week 2** | üîπ **Week 3** | üîπ **Week 4** |
|----------------------|-------------:|-------------:|-------------:|-------------:|
| üì¶ **Unidades Producidas** | **1,124** | **1,310** | **1,092** | **981** |
| üë∑‚Äç‚ôÇÔ∏è **Mano de Obra ($)** | **12,735** | **14,842** | **10,603** | **9,526** |
| üõ†Ô∏è **Materiales ($)** | **21,041** | **24,523** | **20,442** | **18,364** |
| üè≠ **Overhead ($)** | **8,992** | **10,480** | **8,736** | **7,848** |

---

## ‚ùì **Preguntas a Resolver**  

### üî¢ **Parte A: Productividad Multifactorial**  
üìä ¬øLas **mejoras recientes** en el proceso han tenido alg√∫n impacto en la productividad multifactorial?  
üìà ¬øEn qu√© semana se observa el **mayor cambio** en la productividad multifactorial?  

### üë®‚Äçüè≠ **Parte B: Productividad Laboral**  
üìâ ¬øLa **productividad laboral** ha cambiado a lo largo de las semanas?  
üßÆ Utilizar la raz√≥n de **productividad laboral** para justificar la respuesta.  

> üìù **Nota:** Se evaluar√° la **eficiencia del uso de recursos** y el impacto de las mejoras en la producci√≥n.  

---


```r
# ---------------------------------------------
# C√°lculo de Productividad Multifactorial y Laboral
# ---------------------------------------------
# Administraci√≥n de Operaciones: 
# Evaluamos la eficiencia de un proceso mediante las m√©tricas de productividad multifactorial 
# (considerando m√∫ltiples insumos) y productividad laboral (en funci√≥n de horas de trabajo).

# Definimos valores iniciales seg√∫n la imagen
valor_output <- 100  # Precio por unidad producida ($)
costo_laboral <- 50  # Costo por hora de trabajo ($)

# Datos semanales
unidades_producidas <- c(1124, 1310, 1092, 981)
costo_laboral_total <- c(12735, 14842, 10603, 9526)
horas_laborales <- c(254.7, 296.84, 212.06, 190.52)
materiales <- c(21041, 24523, 20442, 18364)
overhead <- c(8992, 10480, 8736, 7848)

# Calculamos el valor total producido
valor_total <- unidades_producidas * valor_output

# C√°lculo de insumo total para productividad multifactorial
insumo_total <- costo_laboral_total + materiales + overhead

# ---------------------------------------------
# Productividad Multifactorial (PMF)
# ---------------------------------------------
productividad_multifactorial <- valor_total / insumo_total

# ---------------------------------------------
# Productividad Laboral (PL)
# ---------------------------------------------
productividad_laboral <- unidades_producidas / horas_laborales

# ---------------------------------------------
# Resultados
# ---------------------------------------------
# Imprimir resultados intermedios
cat("\n--- Resultados de Productividad ---\n")
cat("Semana | Valor Total | Insumo Total | PMF  | Horas | PL  \n")
cat("--------------------------------------------------------\n")
for (i in 1:4) {
  cat(sprintf("  %d    | %10.2f | %12.2f | %.4f | %.2f | %.4f \n",
              i, valor_total[i], insumo_total[i], productividad_multifactorial[i], 
              horas_laborales[i], productividad_laboral[i]))
}

# Determinar la mejora en productividad
mejora_pmf <- ((productividad_multifactorial[4] - productividad_multifactorial[1]) / productividad_multifactorial[1]) * 100
mejora_pl <- ((productividad_laboral[4] - productividad_laboral[1]) / productividad_laboral[1]) * 100

cat("\n--- An√°lisis de Cambio en Productividad ---\n")
cat(sprintf("Cambio en Productividad Multifactorial: %.2f%%\n", mejora_pmf))
cat(sprintf("Cambio en Productividad Laboral: %.2f%%\n", mejora_pl))

# Conclusi√≥n
if (mejora_pmf > 0) {
  cat("\nLa productividad multifactorial ha mejorado a lo largo de las semanas.\n")
} else {
  cat("\nNo hubo mejora significativa en la productividad multifactorial.\n")
}

if (mejora_pl > 0) {
  cat("La productividad laboral tambi√©n ha aumentado, indicando una mayor eficiencia del trabajo.\n")
} else {
  cat("No hubo mejora significativa en la productividad laboral.\n")
}
```

# üè≠ **Ejercicio 07: An√°lisis de Productividad y Utilidades en BBBC**  

---

> üìå **Objetivo:** Evaluar el impacto de un cambio en el **programa de producci√≥n** de **Big Black Bird Company (BBBC)**  
en t√©rminos de **productividad multifactorial, productividad laboral y utilidades semanales**.  

---

## üè¢ **Descripci√≥n del Problema**  

ü¶Ö **Big Black Bird Company (BBBC)** recibi√≥ un **pedido urgente** de uniformes militares con un **revestimiento especial de pl√°stico**.  
üìà Debido a la urgencia, se modific√≥ el programa de producci√≥n, aumentando las horas de trabajo semanales.  

---

## üîπ **Producci√≥n Normal**  

- **üìÜ Turnos:** **2 turnos** de **40 horas** cada uno (por semana).  
- **üëï Producci√≥n:** **2,500 uniformes/semana**.  
- **üí∞ Costo est√°ndar por uniforme:** **$120**.  
- **üë∑ Distribuci√≥n de empleados:**  
  - üè≠ **70 empleados** en el **primer turno**.  
  - üè≠ **30 empleados** en el **segundo turno**.  
- **üìå Precio de contrato:** **$200 por uniforme**.  

---

## üî∫ **Producci√≥n con Turnos Extendidos**  

- üïê **Cada turno trabaja 72 horas por semana**.  
- üìà **Producci√≥n aumenta a 4,000 uniformes/semana**.  
- üí≤ **Costo por uniforme aumenta a $144**.  

---

## ‚ùì **Preguntas a Resolver**  

### üî¢ **Parte A: Productividad Multifactorial**  
üìä ¬øLa **productividad multifactorial** aument√≥, disminuy√≥ o se mantuvo igual?  
üìà Si cambi√≥, ¬øen qu√© **porcentaje**?  

### üë®‚Äçüè≠ **Parte B: Productividad Laboral**  
üìâ ¬øLa **productividad laboral** cambi√≥ con la extensi√≥n de turnos?  
üßÆ Si hubo un cambio, ¬øcu√°l fue el **porcentaje de variaci√≥n**?  

### üí∞ **Parte C: Utilidades Semanales**  
üìà ¬øLas **utilidades semanales** aumentaron, disminuyeron o se mantuvieron iguales tras el cambio de producci√≥n?  

> üìù **Nota:** Se evaluar√° la eficiencia del **uso de recursos, costos y beneficios** asociados con el nuevo programa.  

---



```r
# ------------------------------------------------------------------------------
# C√°lculo de productividad multifactorial, productividad laboral y utilidades
# Administraci√≥n de operaciones - Evaluaci√≥n del impacto en productividad
# ------------------------------------------------------------------------------

# Definimos las variables seg√∫n los datos del problema

# Producci√≥n original
unidades_original <- 2500
costo_produccion_original <- 120
precio_venta <- 200
empleados_turno1 <- 70
empleados_turno2 <- 30
horas_trabajadas_original <- 40 * (empleados_turno1 + empleados_turno2)
costo_total_original <- unidades_original * costo_produccion_original

# Producci√≥n en sobretiempo
unidades_sobrecarga <- 4000
costo_produccion_sobrecarga <- 144
horas_trabajadas_sobrecarga <- 72 * (empleados_turno1 + empleados_turno2)
costo_total_sobrecarga <- unidades_sobrecarga * costo_produccion_sobrecarga

# C√°lculo de productividad multifactorial (PMF)
pmf_original <- unidades_original / costo_total_original
pmf_sobrecarga <- unidades_sobrecarga / costo_total_sobrecarga
cambio_pmf <- ((pmf_sobrecarga - pmf_original) / pmf_original) * 100

# C√°lculo de productividad laboral (PL)
pl_original <- unidades_original / horas_trabajadas_original
pl_sobrecarga <- unidades_sobrecarga / horas_trabajadas_sobrecarga
cambio_pl <- ((pl_sobrecarga - pl_original) / pl_original) * 100

# C√°lculo de utilidades semanales
utilidad_original <- (precio_venta - costo_produccion_original) * unidades_original
utilidad_sobrecarga <- (precio_venta - costo_produccion_sobrecarga) * unidades_sobrecarga
cambio_utilidad <- ((utilidad_sobrecarga - utilidad_original) / utilidad_original) * 100

# ------------------------------------------------------------------------------
# Mostramos los resultados
# ------------------------------------------------------------------------------

cat("**Resultados del an√°lisis de productividad y utilidad**\n\n")

cat("1. Productividad Multifactorial (PMF):\n")
cat(sprintf("   - Productividad original: %.6f\n", pmf_original))
cat(sprintf("   - Productividad con sobrecarga: %.6f\n", pmf_sobrecarga))
cat(sprintf("   - Cambio en PMF: %.2f%%\n\n", cambio_pmf))

cat("2. Productividad Laboral (PL):\n")
cat(sprintf("   - Productividad original: %.6f\n", pl_original))
cat(sprintf("   - Productividad con sobrecarga: %.6f\n", pl_sobrecarga))
cat(sprintf("   - Cambio en PL: %.2f%%\n\n", cambio_pl))

cat("3. Utilidad Semanal:\n")
cat(sprintf("   - Utilidad original: $%.2f\n", utilidad_original))
cat(sprintf("   - Utilidad con sobrecarga: $%.2f\n", utilidad_sobrecarga))
cat(sprintf("   - Cambio en utilidad: %.2f%%\n\n", cambio_utilidad))

# ------------------------------------------------------------------------------
# An√°lisis de los resultados
# ------------------------------------------------------------------------------
cat("Conclusiones:\n")
if (cambio_pmf < 0) {
  cat("   - La productividad multifactorial disminuy√≥ en", round(cambio_pmf, 2), "%.\n")
} else {
  cat("   - La productividad multifactorial aument√≥ en", round(cambio_pmf, 2), "%.\n")
}

if (cambio_pl < 0) {
  cat("   - La productividad laboral disminuy√≥ en", round(cambio_pl, 2), "%.\n")
} else {
  cat("   - La productividad laboral aument√≥ en", round(cambio_pl, 2), "%.\n")
}

if (cambio_utilidad > 0) {
  cat("   - Las utilidades semanales aumentaron en", round(cambio_utilidad, 2), "%.\n")
} else {
  cat("   - Las utilidades semanales disminuyeron en", round(cambio_utilidad, 2), "%.\n")
}

cat("\nInterpretaci√≥n: A pesar de la disminuci√≥n en la eficiencia, el incremento en la producci√≥n permiti√≥ aumentar las utilidades semanales.\n")
```
