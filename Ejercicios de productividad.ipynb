| ğŸ”¢ **Variable**                     | ğŸ“– **DescripciÃ³n**                                      | ğŸ’² **Valor**   |
|--------------------------------------|---------------------------------------------------------|--------------:|
| ğŸ“ **Cuota escolar por hora crÃ©dito**  | Costo que pagan los estudiantes por hora crÃ©dito       | **$150**      |
| ğŸ›ï¸ **Apoyo estatal por hora crÃ©dito**  | Cantidad que aporta el gobierno estatal por hora crÃ©dito | **$100**      |

> ğŸ“ **Nota:** Se debe calcular la productividad teniendo en cuenta todos los costos y beneficios asociados al curso.

```r
# ğŸ“˜ Ejercicio 01 - Productividad en un Curso Universitario

# ğŸ“Š DefiniciÃ³n de variables
# Costos y beneficios del curso
cuota_escolar <- 150  # ğŸ’° Costo por hora crÃ©dito pagado por estudiante (USD)
apoyo_estatal <- 100   # ğŸ›ï¸ Subsidio estatal por hora crÃ©dito (USD)
tamano_grupo <- 50     # ğŸ‘¨â€ğŸ“ NÃºmero de estudiantes por clase
creditos_curso <- 3    # ğŸ“š NÃºmero de crÃ©ditos por curso

# ğŸ·ï¸ Costos asociados
costo_mano_obra <- 4000    # ğŸ§‘â€ğŸ« Costo laboral por clase (USD)
costo_materiales <- 20     # ğŸ“– Costo de materiales por estudiante (USD)
costos_generales <- 25000  # ğŸ¢ Costos generales por clase (USD)

# â³ Horas trabajadas por el instructor
total_semanas <- 16          # ğŸ“… DuraciÃ³n del curso en semanas
horas_por_semana <- 14       # â±ï¸ Horas de trabajo del instructor por semana
horas_trabajadas <- horas_por_semana * total_semanas  # â³ Total de horas trabajadas

# ğŸ’¡ CÃ¡lculo del valor del producto (Outputs)
total_cuota_escolar <- cuota_escolar * creditos_curso * tamano_grupo  # Ingresos por cuota escolar
apoyo_estatal_total <- apoyo_estatal * creditos_curso * tamano_grupo  # Ingresos por subsidio estatal
valor_producto_total <- total_cuota_escolar + apoyo_estatal_total  # ğŸ’µ Ingresos totales del curso

# ğŸ’¡ CÃ¡lculo del valor de los insumos (Inputs)
costo_materiales_total <- costo_materiales * tamano_grupo  # ğŸ“– Costo total de materiales
valor_insumos_total <- costo_mano_obra + costo_materiales_total + costos_generales  # ğŸ’° Costos totales

# ğŸ“ˆ CÃ¡lculo de la productividad multifactorial
# ğŸ”¹ Productividad multifactorial = Valor del Producto / Valor de los Insumos
productividad_multifactorial <- valor_producto_total / valor_insumos_total

# ğŸ“ˆ CÃ¡lculo de la productividad laboral
# ğŸ”¹ Productividad laboral = Valor del Producto / Horas trabajadas
productividad_laboral <- valor_producto_total / horas_trabajadas

# ğŸ“¢ ImpresiÃ³n de resultados
cat("=== ğŸ“Š Resultados del Ejercicio ===\n\n")

cat("ğŸ“Œ Valor del Producto (Outputs):\n")
cat("ğŸ“ Cuota Escolar Total: $", total_cuota_escolar, "\n")
cat("ğŸ›ï¸ Apoyo Estatal Total: $", apoyo_estatal_total, "\n")
cat("ğŸ’µ Valor Total del Producto: $", valor_producto_total, "\n\n")

cat("ğŸ“Œ Valor de los Insumos (Inputs):\n")
cat("ğŸ§‘â€ğŸ« Costo de Mano de Obra: $", costo_mano_obra, "\n")
cat("ğŸ“– Costo de Materiales Total: $", costo_materiales_total, "\n")
cat("ğŸ¢ Costos Generales: $", costos_generales, "\n")
cat("ğŸ’° Valor Total de los Insumos: $", valor_insumos_total, "\n\n")

cat("ğŸ“Œ Productividad Multifactorial:\n")
cat("âš–ï¸ Productividad = Valor del Producto / Valor de los Insumos =", productividad_multifactorial, "\n\n")

cat("ğŸ“Œ Productividad Laboral:\n")
cat("â³ Horas trabajadas por instructor: ", horas_trabajadas, " horas\n")
cat("ğŸ’µ Ingresos por curso: $", valor_producto_total, "\n")
cat("ğŸ“Š Productividad Laboral (USD por hora):", productividad_laboral, "\n")
```

```r
# ğŸ“Œ **Ejercicio 05: OptimizaciÃ³n de Costos en White Tiger Electronics**
# =========================================================
# ğŸ”§ CÃ¡lculo de reducciÃ³n de costos para mejorar productividad multifactorial
# ğŸ“Š AdministraciÃ³n de operaciones - AnÃ¡lisis de eficiencia en costos
# =========================================================

# ğŸ“Œ DefiniciÃ³n de variables clave
precio_venta <- 300       # ğŸ’° Precio de venta por unidad ($)
costo_total <- 150        # ğŸ·ï¸ Costo total de producciÃ³n por unidad ($)
costo_mano_obra <- 30     # ğŸ‘· Costo de mano de obra por unidad ($)
costo_materiales <- 70    # ğŸ› ï¸ Costo de materiales por unidad ($)
costo_overhead <- 50      # ğŸ­ Costos generales (Overhead) por unidad ($)
mejora_productividad <- 1.1  # ğŸ“ˆ Mejora esperada del 10% en productividad multifactorial

# ğŸ“Š CÃ¡lculo de la nueva entrada de recursos requerida
nuevo_input <- costo_total / mejora_productividad

# ğŸ“‰ CÃ¡lculo de la reducciÃ³n total necesaria en costos
reduccion_necesaria <- costo_total - nuevo_input

# ğŸ”¢ CÃ¡lculo de reducciÃ³n en Materiales
reduccion_materiales <- (reduccion_necesaria / costo_materiales) * 100
nuevo_costo_materiales <- costo_materiales - reduccion_necesaria

# ğŸ”¢ CÃ¡lculo de reducciÃ³n en Mano de Obra
reduccion_mano_obra <- (reduccion_necesaria / costo_mano_obra) * 100
nuevo_costo_mano_obra <- costo_mano_obra - reduccion_necesaria

# ğŸ”¢ CÃ¡lculo de reducciÃ³n en Costos Generales (Overhead)
reduccion_overhead <- (reduccion_necesaria / costo_overhead) * 100
nuevo_costo_overhead <- costo_overhead - reduccion_necesaria

# ğŸ“‹ ImpresiÃ³n de resultados
cat("=======================================\n")
cat(" White Tiger Electronics - Resultados \n")
cat("=======================================\n\n")

cat("ğŸ·ï¸ Costo total original de producciÃ³n por unidad:", costo_total, "\n")
cat("ğŸ“Š Nuevo costo total necesario para lograr mejora del 10%:", round(nuevo_input, 4), "\n")
cat("ğŸ“‰ ReducciÃ³n total requerida en costos:", round(reduccion_necesaria, 4), "\n\n")

cat("ğŸ› ï¸ ReducciÃ³n necesaria en Materiales:\n")
cat(" - Nuevo costo de materiales:", round(nuevo_costo_materiales, 4), "\n")
cat(" - Porcentaje de reducciÃ³n:", round(reduccion_materiales, 2), "%\n\n")

cat("ğŸ‘· ReducciÃ³n necesaria en Mano de Obra:\n")
cat(" - Nuevo costo de mano de obra:", round(nuevo_costo_mano_obra, 4), "\n")
cat(" - Porcentaje de reducciÃ³n:", round(reduccion_mano_obra, 2), "%\n\n")

cat("ğŸ­ ReducciÃ³n necesaria en Costos Generales (Overhead):\n")
cat(" - Nuevo costo de overhead:", round(nuevo_costo_overhead, 4), "\n")
cat(" - Porcentaje de reducciÃ³n:", round(reduccion_overhead, 2), "%\n")
```

## ğŸ¢ **DescripciÃ³n del Proceso**  

âš™ï¸ En **Symtecks**, un proceso especÃ­fico genera un **output** con un valor de **$100 por unidad producida**.  
ğŸ‘·â€â™‚ï¸ El **costo de la mano de obra** es de **$50 por hora**, incluyendo beneficios.  

ğŸ“Š **El departamento de contabilidad** ha proporcionado informaciÃ³n sobre la producciÃ³n y costos en las Ãºltimas **cuatro semanas**.

## ğŸ“Š **Datos del Proceso**  

| ğŸ“… **Concepto**       | ğŸ”¹ **Week 1** | ğŸ”¹ **Week 2** | ğŸ”¹ **Week 3** | ğŸ”¹ **Week 4** |
|----------------------|-------------:|-------------:|-------------:|-------------:|
| ğŸ“¦ **Unidades Producidas** | **1,124** | **1,310** | **1,092** | **981** |
| ğŸ‘·â€â™‚ï¸ **Mano de Obra ($)** | **12,735** | **14,842** | **10,603** | **9,526** |
| ğŸ› ï¸ **Materiales ($)** | **21,041** | **24,523** | **20,442** | **18,364** |
| ğŸ­ **Overhead ($)** | **8,992** | **10,480** | **8,736** | **7,848** |

---

## â“ **Preguntas a Resolver**  

### ğŸ”¢ **Parte A: Productividad Multifactorial**  
ğŸ“Š Â¿Las **mejoras recientes** en el proceso han tenido algÃºn impacto en la productividad multifactorial?  
ğŸ“ˆ Â¿En quÃ© semana se observa el **mayor cambio** en la productividad multifactorial?  

### ğŸ‘¨â€ğŸ­ **Parte B: Productividad Laboral**  
ğŸ“‰ Â¿La **productividad laboral** ha cambiado a lo largo de las semanas?  
ğŸ§® Utilizar la razÃ³n de **productividad laboral** para justificar la respuesta.  

> ğŸ“ **Nota:** Se evaluarÃ¡ la **eficiencia del uso de recursos** y el impacto de las mejoras en la producciÃ³n.  

---


```r
# ğŸ­ **CÃ¡lculo de Productividad Multifactorial y Laboral**
# -------------------------------------------------------
# ğŸ“Œ Evaluamos la eficiencia del proceso en Symtecks mediante dos mÃ©tricas clave:
#    1ï¸âƒ£ Productividad Multifactorial (PMF): RelaciÃ³n entre el valor de producciÃ³n y el costo total de insumos.
#    2ï¸âƒ£ Productividad Laboral (PL): RelaciÃ³n entre unidades producidas y horas trabajadas.

# ğŸ“Š **Datos Iniciales**
valor_output <- 100  # Precio por unidad producida ($)
costo_laboral <- 50   # Costo por hora de trabajo ($)

# ğŸ”¹ **Datos Semanales**
unidades_producidas <- c(1124, 1310, 1092, 981)
costo_laboral_total <- c(12735, 14842, 10603, 9526)
horas_laborales <- c(254.7, 296.84, 212.06, 190.52)
materiales <- c(21041, 24523, 20442, 18364)
overhead <- c(8992, 10480, 8736, 7848)

# ğŸ” **CÃ¡lculos**
# -------------------------------------------------------
# ğŸ“Œ **Paso 1: Calcular el valor total de producciÃ³n**
valor_total <- unidades_producidas * valor_output

# ğŸ“Œ **Paso 2: Calcular el insumo total** (suma de costos laborales, materiales y overhead)
insumo_total <- costo_laboral_total + materiales + overhead

# ğŸ“Œ **Paso 3: Calcular la Productividad Multifactorial (PMF)**
productividad_multifactorial <- valor_total / insumo_total

# ğŸ“Œ **Paso 4: Calcular la Productividad Laboral (PL)**
productividad_laboral <- unidades_producidas / horas_laborales

# ğŸ“Š **Resultados**
# -------------------------------------------------------
cat("\n--- ğŸ“ˆ Resultados de Productividad ---\n")
cat("Semana | Valor Total | Insumo Total | PMF    | Horas  | PL    \n")
cat("----------------------------------------------------------\n")
for (i in 1:4) {
  cat(sprintf("  %d    | %10.2f | %12.2f | %.4f | %.2f | %.4f \n",
              i, valor_total[i], insumo_total[i], productividad_multifactorial[i], 
              horas_laborales[i], productividad_laboral[i]))
}

# ğŸ” **AnÃ¡lisis de Mejora**
# -------------------------------------------------------
mejora_pmf <- ((productividad_multifactorial[4] - productividad_multifactorial[1]) / productividad_multifactorial[1]) * 100
mejora_pl <- ((productividad_laboral[4] - productividad_laboral[1]) / productividad_laboral[1]) * 100

cat("\n--- ğŸ“Š AnÃ¡lisis de Cambio en Productividad ---\n")
cat(sprintf("ğŸ“Œ Cambio en Productividad Multifactorial: %.2f%%\n", mejora_pmf))
cat(sprintf("ğŸ“Œ Cambio en Productividad Laboral: %.2f%%\n", mejora_pl))

# ğŸ **ConclusiÃ³n**
# -------------------------------------------------------
if (mejora_pmf > 0) {
  cat("\nâœ… La productividad multifactorial ha mejorado a lo largo de las semanas.\n")
} else {
  cat("\nâŒ No hubo mejora significativa en la productividad multifactorial.\n")
}

if (mejora_pl > 0) {
  cat("âœ… La productividad laboral tambiÃ©n ha aumentado, indicando una mayor eficiencia del trabajo.\n")
} else {
  cat("âŒ No hubo mejora significativa en la productividad laboral.\n")
}
```

# ğŸ­ **Ejercicio 07: AnÃ¡lisis de Productividad y Utilidades en BBBC**  

---

> ğŸ“Œ **Objetivo:** Evaluar el impacto de un cambio en el **programa de producciÃ³n** de **Big Black Bird Company (BBBC)**  
en tÃ©rminos de **productividad multifactorial, productividad laboral y utilidades semanales**.

## ğŸ¢ **DescripciÃ³n del Problema**  

ğŸ¦… **Big Black Bird Company (BBBC)** recibiÃ³ un **pedido urgente** de uniformes militares con un **revestimiento especial de plÃ¡stico**.  
ğŸ“ˆ Debido a la urgencia, se modificÃ³ el programa de producciÃ³n, aumentando las horas de trabajo semanales.  

---

## ğŸ”¹ **ProducciÃ³n Normal**  

- **ğŸ“† Turnos:** **2 turnos** de **40 horas** cada uno (por semana).  
- **ğŸ‘• ProducciÃ³n:** **2,500 uniformes/semana**.  
- **ğŸ’° Costo estÃ¡ndar por uniforme:** **$120**.  
- **ğŸ‘· DistribuciÃ³n de empleados:**  
  - ğŸ­ **70 empleados** en el **primer turno**.  
  - ğŸ­ **30 empleados** en el **segundo turno**.  
- **ğŸ“Œ Precio de contrato:** **$200 por uniforme**.  

---

## ğŸ”º **ProducciÃ³n con Turnos Extendidos**  

- ğŸ• **Cada turno trabaja 72 horas por semana**.  
- ğŸ“ˆ **ProducciÃ³n aumenta a 4,000 uniformes/semana**.  
- ğŸ’² **Costo por uniforme aumenta a $144**.  

---

## â“ **Preguntas a Resolver**  

### ğŸ”¢ **Parte A: Productividad Multifactorial**  
ğŸ“Š Â¿La **productividad multifactorial** aumentÃ³, disminuyÃ³ o se mantuvo igual?  
ğŸ“ˆ Si cambiÃ³, Â¿en quÃ© **porcentaje**?  

### ğŸ‘¨â€ğŸ­ **Parte B: Productividad Laboral**  
ğŸ“‰ Â¿La **productividad laboral** cambiÃ³ con la extensiÃ³n de turnos?  
ğŸ§® Si hubo un cambio, Â¿cuÃ¡l fue el **porcentaje de variaciÃ³n**?  

### ğŸ’° **Parte C: Utilidades Semanales**  
ğŸ“ˆ Â¿Las **utilidades semanales** aumentaron, disminuyeron o se mantuvieron iguales tras el cambio de producciÃ³n?  

> ğŸ“ **Nota:** Se evaluarÃ¡ la eficiencia del **uso de recursos, costos y beneficios** asociados con el nuevo programa.  

---



```r
# ğŸ­ **AnÃ¡lisis de Productividad y Utilidades en BBBC**
# -----------------------------------------------------
# ğŸ“Œ Evaluamos el impacto del cambio en el programa de producciÃ³n de BBBC
#    en tÃ©rminos de productividad multifactorial, productividad laboral y utilidades semanales.

# -----------------------------------------------------
# ğŸ“Š Definimos las variables segÃºn los datos del problema
# -----------------------------------------------------

# ğŸ”¹ ProducciÃ³n Normal
unidades_original <- 2500  # Uniformes producidos por semana
costo_produccion_original <- 120  # Costo por uniforme ($)
precio_venta <- 200  # Precio de venta por uniforme ($)

# ğŸ‘· DistribuciÃ³n de empleados y horas trabajadas
empleados_turno1 <- 70  # Empleados en el turno 1
empleados_turno2 <- 30  # Empleados en el turno 2
horas_trabajadas_original <- 40 * (empleados_turno1 + empleados_turno2)  # Horas totales trabajadas

# ğŸ’° Costos Totales
costo_total_original <- unidades_original * costo_produccion_original  # Costo total de producciÃ³n ($)

# ğŸ”º ProducciÃ³n con Turnos Extendidos
unidades_sobrecarga <- 4000  # Uniformes producidos por semana con turnos extendidos
costo_produccion_sobrecarga <- 144  # Nuevo costo por uniforme ($)
horas_trabajadas_sobrecarga <- 72 * (empleados_turno1 + empleados_turno2)  # Horas totales trabajadas en sobrecarga
costo_total_sobrecarga <- unidades_sobrecarga * costo_produccion_sobrecarga  # Costo total de producciÃ³n ($)

# -----------------------------------------------------
# ğŸ“Š CÃ¡lculo de Productividad Multifactorial (PMF)
# -----------------------------------------------------
pmf_original <- unidades_original / costo_total_original  # Productividad original
pmf_sobrecarga <- unidades_sobrecarga / costo_total_sobrecarga  # Productividad con turnos extendidos
cambio_pmf <- ((pmf_sobrecarga - pmf_original) / pmf_original) * 100  # Cambio porcentual en PMF

# -----------------------------------------------------
# ğŸ‘¨â€ğŸ­ CÃ¡lculo de Productividad Laboral (PL)
# -----------------------------------------------------
pl_original <- unidades_original / horas_trabajadas_original  # Productividad laboral original
pl_sobrecarga <- unidades_sobrecarga / horas_trabajadas_sobrecarga  # Productividad laboral con sobrecarga
cambio_pl <- ((pl_sobrecarga - pl_original) / pl_original) * 100  # Cambio porcentual en PL

# -----------------------------------------------------
# ğŸ’° CÃ¡lculo de Utilidades Semanales
# -----------------------------------------------------
utilidad_original <- (precio_venta - costo_produccion_original) * unidades_original  # Utilidad semanal original ($)
utilidad_sobrecarga <- (precio_venta - costo_produccion_sobrecarga) * unidades_sobrecarga  # Utilidad semanal con sobrecarga ($)
cambio_utilidad <- ((utilidad_sobrecarga - utilidad_original) / utilidad_original) * 100  # Cambio porcentual en utilidad

# -----------------------------------------------------
# ğŸ“¢ Mostramos los resultados
# -----------------------------------------------------
cat("ğŸ­ **Resultados del AnÃ¡lisis de Productividad y Utilidad**\n\n")

cat("ğŸ“Š 1. Productividad Multifactorial (PMF):\n")
cat(sprintf("   - Productividad original: %.6f\n", pmf_original))
cat(sprintf("   - Productividad con sobrecarga: %.6f\n", pmf_sobrecarga))
cat(sprintf("   - Cambio en PMF: %.2f%%\n\n", cambio_pmf))

cat("ğŸ‘¨â€ğŸ­ 2. Productividad Laboral (PL):\n")
cat(sprintf("   - Productividad original: %.6f\n", pl_original))
cat(sprintf("   - Productividad con sobrecarga: %.6f\n", pl_sobrecarga))
cat(sprintf("   - Cambio en PL: %.2f%%\n\n", cambio_pl))

cat("ğŸ’° 3. Utilidad Semanal:\n")
cat(sprintf("   - Utilidad original: $%.2f\n", utilidad_original))
cat(sprintf("   - Utilidad con sobrecarga: $%.2f\n", utilidad_sobrecarga))
cat(sprintf("   - Cambio en utilidad: %.2f%%\n\n", cambio_utilidad))

# -----------------------------------------------------
# ğŸ” AnÃ¡lisis de los Resultados
# -----------------------------------------------------
cat("ğŸ“Œ **Conclusiones:**\n")
if (cambio_pmf < 0) {
  cat("   - La productividad multifactorial disminuyÃ³ en", round(cambio_pmf, 2), "% ğŸš¨.\n")
} else {
  cat("   - La productividad multifactorial aumentÃ³ en", round(cambio_pmf, 2), "% âœ….\n")
}

if (cambio_pl < 0) {
  cat("   - La productividad laboral disminuyÃ³ en", round(cambio_pl, 2), "% ğŸš¨.\n")
} else {
  cat("   - La productividad laboral aumentÃ³ en", round(cambio_pl, 2), "% âœ….\n")
}

if (cambio_utilidad > 0) {
  cat("   - Las utilidades semanales aumentaron en", round(cambio_utilidad, 2), "% âœ….\n")
} else {
  cat("   - Las utilidades semanales disminuyeron en", round(cambio_utilidad, 2), "% ğŸš¨.\n")
}

cat("\nğŸ“Š **InterpretaciÃ³n:** A pesar de la reducciÃ³n en la eficiencia, el aumento en la producciÃ³n permitiÃ³ incrementar las utilidades semanales.\n")
```
