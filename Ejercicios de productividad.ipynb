**Ejercicio01**

---

Calcular y analizar la productividad multifactorial y la productividad laboral en un curso típico de 3 créditos en la Universidad *Boehring*  

Student tuition at *Boehring* University is $150 per semester credit hour.  
The state supplements school revenue by $100 per semester credit hour.  
Average class size for a typical 3-credit course is 50 students.  

Los costos asociados son:  

- **Labor costs**: $4,000 por clase  
- **Material costs**: $20 por estudiante por clase  
- **Overhead costs**: $25,000 por clase  

**Problemas a resolver**

a. ¿Cuál es la razón de productividad *multifactorial* para este proceso de curso?  
b. Si los instructores trabajan un promedio de 14 horas por semana durante 16 semanas para cada clase de 3 créditos con 50 estudiantes, ¿cuál es la razón de productividad *laboral*?  

---

**Tabla 1: Datos del problema**

| **Variable**                        | **Descripción**                                         | **Valor**   |
|--------------------------------------|---------------------------------------------------------|------------:|
| Cuota escolar por hora crédito       | Costo que pagan los estudiantes por hora crédito       | $150        |
| Apoyo estatal por hora crédito       | Cantidad que aporta el gobierno estatal por hora crédito | $100        |
| Tamaño promedio del grupo            | Número de estudiantes por grupo                        | 50          |
| Número de créditos por curso         | Número de créditos por curso                           | 3           |

---



```r
# Definición de variables
cuota_escolar <- 150  # USD por hora crédito
apoyo_estatal <- 100   # USD por hora crédito
tamano_grupo <- 50     # Número de estudiantes
creditos_curso <- 3    # Número de créditos por curso

costo_mano_obra <- 4000    # USD por clase
costo_materiales <- 20     # USD por estudiante
costos_generales <- 25000  # USD por clase
horas_trabajadas <- 14 * 16  # 14 horas/semana por 16 semanas

# Cálculo del valor del producto (OUTPUTS)
cuota_escolar_total <- cuota_escolar * creditos_curso * tamano_grupo
apoyo_estatal_total <- apoyo_estatal * creditos_curso * tamano_grupo
valor_producto_total <- cuota_escolar_total + apoyo_estatal_total

# Cálculo del valor de los insumos (INPUTS)
costo_materiales_total <- costo_materiales * tamano_grupo
valor_insumos_total <- costo_mano_obra + costo_materiales_total + costos_generales

# Cálculo de la productividad multifactorial
productividad_multifactorial <- valor_producto_total / valor_insumos_total

# Cálculo de la productividad laboral
productividad_laboral <- valor_producto_total / horas_trabajadas

# Impresión de resultados
cat("=== Resultados del Ejercicio ===\n\n")

cat("Valor del Producto (Outputs):\n")
cat("Cuota Escolar Total: $", cuota_escolar_total, "\n")
cat("Apoyo Estatal Total: $", apoyo_estatal_total, "\n")
cat("Valor Total del Producto: $", valor_producto_total, "\n\n")

cat("Valor de los Insumos (Inputs):\n")
cat("Costo de Mano de Obra: $", costo_mano_obra, "\n")
cat("Costo de Materiales Total: $", costo_materiales_total, "\n")
cat("Costos Generales: $", costos_generales, "\n")
cat("Valor Total de los Insumos: $", valor_insumos_total, "\n\n")

cat("Productividad Multifactorial:\n")
cat("Productividad = Valor del Producto / Valor de los Insumos = ", productividad_multifactorial, "\n\n")

cat("Productividad Laboral:\n")
cat("Horas trabajadas por maestro: ", horas_trabajadas, " horas\n")
cat("Ingresos por curso: $", valor_producto_total, "\n")
cat("Productividad Laboral (USD por hora): ", productividad_laboral, "\n")
```

**Ejercicio02**

---

Calcular y analizar la productividad laboral en la producción de prendas de vestir en *Natalie Attire*  

Natalie Attire fabrica prendas de moda.  
Durante una semana en particular, los empleados trabajaron 360 horas para producir un lote de 132 prendas, de las cuales 52 eran de "segunda" (con defectos).  

- Las prendas de segunda se venden a **$90** cada una en la tienda de fábrica de Attire.  
- Las **80 prendas restantes** se venden a **$200** cada una en la distribución minorista.  

**Problema a resolver**
¿Qué razón de productividad laboral (*labor productivity ratio*) tiene este proceso de manufactura?  

---

**Tabla 1: Datos del problema**

| **Variable**                        | **Valor**   |
|--------------------------------------|------------:|
| Número total de prendas              | 132        |
| Número de prendas de segunda         | 52         |
| Número de prendas restantes          | 80         |
| Precio de venta por prenda de segunda | $90        |
| Precio de venta por prenda restante  | $200       |
| Horas de trabajo empleadas           | 360        |

---



```r
# Definir variables
num_prendas_totales <- 132
num_prendas_segunda <- 52
num_prendas_buenas <- 80
precio_segunda <- 90
precio_buenas <- 200
horas_trabajo <- 360

# Cálculo del valor total del producto
valor_prendas_segunda <- num_prendas_segunda * precio_segunda
valor_prendas_buenas <- num_prendas_buenas * precio_buenas
valor_total_producto <- valor_prendas_segunda + valor_prendas_buenas

# Cálculo de la productividad laboral
productividad_laboral <- valor_total_producto / horas_trabajo

# Mostrar resultados
cat("---- Resultados ----\n")
cat("Valor total de prendas de segunda: $", valor_prendas_segunda, "\n")
cat("Valor total de prendas buenas: $", valor_prendas_buenas, "\n")
cat("Valor total del producto: $", valor_total_producto, "\n")
cat("Horas de trabajo empleadas:", horas_trabajo, "horas\n")
cat("Productividad laboral: $", round(productividad_laboral, 2), "por hora\n")
```

---

Analizar la productividad multifactorial y laboral en un escenario de cambios en la Universidad *Boehring*

---

**Trends and Challenges in Operations Management**

(Refer to Solved Problem 1.) Coach **Bijoum** Toulouse led the Big Red Herrings to several disappointing football seasons.  
Only better recruiting will return the Big Red Herrings to winning form.  

Because of the current state of the program, **Boehring** University fans are unlikely to support increases in the $192 season ticket price.  

Improved recruitment will increase overhead costs to $30,000 per class section from the current $25,000 per class section.  

The university’s budget plan is to cover recruitment costs by increasing the average class size to 75 students.  

Labor costs will increase to $6,500 per 3-credit course.  

Material costs will be about $25 per student for each 3-credit course.  

Tuition will be $200 per semester credit, which is supplemented by state support of $100 per semester credit.  

---

**Preguntas:**

**a.** ¿Cuál es el índice de productividad multifactorial? En comparación con el resultado obtenido en el Problema Resuelto 1, ¿la productividad aumentó o disminuyó para el proceso del curso?  

**b.** Si los instructores trabajan un promedio de 20 horas por semana durante 16 semanas para cada curso de 3 créditos con 75 estudiantes, ¿cuál es el índice de productividad laboral?  

---

**Tabla 1: Datos del problema**

| Variable                           | Valor    |
|-------------------------------------|---------:|
| Costo general por sección de clase (nuevo) | 30,000  |
| Costo general por sección de clase (actual) | 25,000  |
| Tamaño promedio de la clase        | 75       |
| Costos laborales por curso         | 6,500    |
| Costos de materiales por estudiante | 25      |
| Costo de matrícula por crédito     | 200      |
| Apoyo estatal por crédito          | 100      |
| Créditos por curso                 | 3        |
| Horas de trabajo del instructor    | 320      |

---


```r
# ========================================================
# Cálculo de Productividad Laboral y Multifactorial en R
# ========================================================
# Administración de Operaciones: La productividad mide la eficiencia 
# en el uso de recursos para generar productos o servicios.
# En este ejercicio se analizan cambios en los costos y horas trabajadas,
# evaluando el impacto en la productividad.

# Definición de variables iniciales
# ----------------------------------

# Costos y parámetros originales
cuota_escolar_original <- 150
apoyo_estatal_original <- 100
tamano_grupo_original <- 50
creditos_curso <- 3
costo_mano_obra_original <- 4000
costo_generales_original <- 25000
costo_materiales_original <- 1000
horas_trabajo_original <- 224

# Costos y parámetros ajustados
cuota_escolar_ajuste <- 200
apoyo_estatal_ajuste <- 100
tamano_grupo_ajuste <- 75
costo_mano_obra_ajuste <- 6500
costo_generales_ajuste <- 30000
costo_materiales_ajuste <- 1875
horas_trabajo_ajuste <- 320

# Cálculo del valor del producto (Ingresos generados)
# ----------------------------------------------------
valor_producto_original <- (cuota_escolar_original + apoyo_estatal_original) * creditos_curso * tamano_grupo_original
valor_producto_ajuste <- (cuota_escolar_ajuste + apoyo_estatal_ajuste) * creditos_curso * tamano_grupo_ajuste

# Cálculo de los insumos (Costos totales)
# ---------------------------------------
insumos_original <- costo_mano_obra_original + costo_generales_original + costo_materiales_original
insumos_ajuste <- costo_mano_obra_ajuste + costo_generales_ajuste + costo_materiales_ajuste

# Cálculo del índice de productividad multifactorial
# --------------------------------------------------
productividad_multifactorial_original <- valor_producto_original / insumos_original
productividad_multifactorial_ajuste <- valor_producto_ajuste / insumos_ajuste

# Cálculo de la productividad laboral (por hora trabajada)
# --------------------------------------------------------
productividad_laboral_original <- valor_producto_original / horas_trabajo_original
productividad_laboral_ajuste <- valor_producto_ajuste / horas_trabajo_ajuste

# Cálculo del cambio porcentual en la productividad
# -------------------------------------------------
cambio_productividad_multifactorial <- ((productividad_multifactorial_ajuste - productividad_multifactorial_original) / productividad_multifactorial_original) * 100
cambio_productividad_laboral <- ((productividad_laboral_ajuste - productividad_laboral_original) / productividad_laboral_original) * 100

# Resultados
# ==========
cat("\n--- Resultados de Productividad ---\n")
cat("Valor del Producto (Original):", valor_producto_original, "\n")
cat("Valor del Producto (Ajustado):", valor_producto_ajuste, "\n")
cat("Productividad Multifactorial Original:", round(productividad_multifactorial_original, 4), "\n")
cat("Productividad Multifactorial Ajustada:", round(productividad_multifactorial_ajuste, 4), "\n")
cat("Cambio en Productividad Multifactorial:", round(cambio_productividad_multifactorial, 2), "%\n")
cat("Productividad Laboral Original:", round(productividad_laboral_original, 4), "\n")
cat("Productividad Laboral Ajustada:", round(productividad_laboral_ajuste, 4), "\n")
cat("Cambio en Productividad Laboral:", round(cambio_productividad_laboral, 2), "%\n")

# ========================================================
# CONCLUSIÓN:
# - Se observa un incremento en la productividad laboral y multifactorial.
# - El aumento de costos fue compensado por un mayor tamaño de grupo y eficiencia.
# - Es crucial evaluar continuamente estos indicadores para optimizar recursos.
# ========================================================
```

Analizar la productividad laboral en el servicio de lavandería *Suds and Duds Laundry*  

Suds and Duds Laundry washed and pressed the following numbers of dress shirts per week.

---

**Datos de Producción**

| Semana | Trabajadores     | Total de horas | Camisas |
|--------|-----------------|---------------:|--------:|
| 1      | Sud y Dud       |             24 |      68 |
| 2      | Sud y Dud       |             46 |     130 |
| 3      | Sud, Dud y Jud  |             62 |     152 |
| 4      | Sud, Dud y Jud  |             51 |     125 |
| 5      | Dud y Jud       |             45 |     131 |

---

**Preguntas**

**a.** Calcular la razón de productividad laboral para cada semana.  

**b.** Explicar el patrón de productividad laboral exhibido por los datos.  


```r
# -----------------------------------------------------------
# Cálculo de la Productividad Laboral - Suds and Duds Laundry
# -----------------------------------------------------------
# La productividad laboral se calcula como:
# Productividad = Producción (output) / Horas de trabajo (input)

# Definir los datos del problema
semanas <- c(1, 2, 3, 4, 5)  # Número de semanas
trabajadores <- c(2, 2, 3, 3, 2)  # Número de trabajadores por semana
horas_trabajo <- c(24, 46, 62, 51, 45)  # Total de horas trabajadas por semana
produccion <- c(68, 130, 152, 125, 131)  # Camisas producidas por semana

# Calcular la productividad laboral (camisas por hora)
productividad <- produccion / horas_trabajo  

# Imprimir los resultados de productividad laboral
cat("Productividad Laboral (camisas por hora):\n")
for (i in 1:length(semanas)) {
  cat(sprintf("Semana %d: %.4f camisas/hora\n", semanas[i], productividad[i]))
}

# Analizar los cambios en productividad entre semanas
cambio_productividad <- c(NA, diff(productividad) / productividad[-length(productividad)] * 100)

# Imprimir los cambios porcentuales de productividad
cat("\nCambio porcentual en productividad entre semanas:\n")
for (i in 2:length(semanas)) {
  cat(sprintf("Semana %d vs Semana %d: %.2f%%\n", semanas[i], semanas[i-1], cambio_productividad[i]))
}

# Análisis de productividad laboral
cat("\nAnálisis de la productividad laboral:\n")
cat("La productividad laboral disminuye en la Semana 3 cuando hay 3 trabajadores.\n")
cat("Esto sugiere que el tercer trabajador no contribuye significativamente al rendimiento.\n")
cat("Podría deberse a restricciones de espacio, equipos o flujo de trabajo ineficiente.\n")
```

White Tiger Electronics produce reproductores de CD utilizando un proceso de ensamblaje automatizado.

El costo estándar de los reproductores de CD es de **$150 por unidad**, con la siguiente distribución de costos:

El precio de venta es de **$300 por unidad**.

**Preguntas**

**b.** Para lograr una mejora del **10%** en la productividad multifactorial reduciendo **únicamente** los costos de mano de obra, ¿en qué porcentaje deben reducirse estos costos?

**c.** Para lograr una mejora del **10%** en la productividad multifactorial reduciendo **únicamente** los costos generales, ¿en qué porcentaje deben reducirse estos costos?


```r
# =========================================================
# Cálculo de reducción de costos para mejorar productividad multifactorial
# Administración de operaciones - Análisis de eficiencia en costos
# =========================================================

# Definición de variables
precio_venta <- 300       # Precio de venta por unidad ($)
costo_total <- 150        # Costo total de producción por unidad ($)
costo_mano_obra <- 30     # Costo de mano de obra por unidad ($)
costo_materiales <- 70    # Costo de materiales por unidad ($)
costo_overhead <- 50      # Costo de costos generales por unidad ($)
mejora_productividad <- 1.1  # Mejora del 10% en la productividad multifactorial

# Cálculo de la nueva entrada de recursos requerida
nuevo_input <- costo_total / mejora_productividad

# Cálculo de la reducción necesaria en costos
reduccion_necesaria <- costo_total - nuevo_input

# Cálculo de reducción en Materiales
reduccion_materiales <- (reduccion_necesaria / costo_materiales) * 100
nuevo_costo_materiales <- costo_materiales - reduccion_necesaria

# Cálculo de reducción en Mano de Obra
reduccion_mano_obra <- (reduccion_necesaria / costo_mano_obra) * 100
nuevo_costo_mano_obra <- costo_mano_obra - reduccion_necesaria

# Cálculo de reducción en Overhead
reduccion_overhead <- (reduccion_necesaria / costo_overhead) * 100
nuevo_costo_overhead <- costo_overhead - reduccion_necesaria

# Impresión de resultados
cat("=======================================\n")
cat(" White Tiger Electronics - Resultados \n")
cat("=======================================\n\n")

cat("Costo total original de producción por unidad:", costo_total, "\n")
cat("Nuevo costo total necesario para lograr mejora del 10%:", round(nuevo_input, 4), "\n")
cat("Reducción total requerida en costos:", round(reduccion_necesaria, 4), "\n\n")

cat("Reducción necesaria en Materiales:\n")
cat(" - Nuevo costo de materiales:", round(nuevo_costo_materiales, 4), "\n")
cat(" - Porcentaje de reducción:", round(reduccion_materiales, 2), "%\n\n")

cat("Reducción necesaria en Mano de Obra:\n")
cat(" - Nuevo costo de mano de obra:", round(nuevo_costo_mano_obra, 4), "\n")
cat(" - Porcentaje de reducción:", round(reduccion_mano_obra, 2), "%\n\n")

cat("Reducción necesaria en Costos Generales (Overhead):\n")
cat(" - Nuevo costo de overhead:", round(nuevo_costo_overhead, 4), "\n")
cat(" - Porcentaje de reducción:", round(reduccion_overhead, 2), "%\n")
```

**Ejercicio06**

---

Problema06: Analizar el impacto de las mejoras recientes en un proceso específico de **Symtecks**, utilizando la productividad multifactorial y la productividad laboral como medidas de desempeño.

En **Symtecks**, el output de un proceso específico tiene un valor de **$100 por unidad**.  
El costo de la mano de obra es **$50 por hora**, incluyendo beneficios.  

El departamento de contabilidad proporcionó la siguiente información sobre el proceso durante las últimas cuatro semanas:

---

**Datos del proceso:**

| Concepto       | **Week 1** | **Week 2** | **Week 3** | **Week 4** |
|---------------|----------:|----------:|----------:|----------:|
| **Unidades Producidas** | 1,124 | 1,310 | 1,092 | 981 |
| **Mano de Obra ($)** | 12,735 | 14,842 | 10,603 | 9,526 |
| **Materiales ($)** | 21,041 | 24,523 | 20,442 | 18,364 |
| **Overhead ($)** | 8,992 | 10,480 | 8,736 | 7,848 |

---

**Preguntas:**

**a.** Utilizar la razón de productividad multifactorial para determinar si las mejoras recientes en el proceso han tenido algún efecto.  
Si es así, indicar en qué semana fue más notable.  

**b.** Ha cambiado la productividad laboral?  
Usar la razón de productividad laboral para justificar la respuesta.

---


```r
# ---------------------------------------------
# Cálculo de Productividad Multifactorial y Laboral
# ---------------------------------------------
# Administración de Operaciones: 
# Evaluamos la eficiencia de un proceso mediante las métricas de productividad multifactorial 
# (considerando múltiples insumos) y productividad laboral (en función de horas de trabajo).

# Definimos valores iniciales según la imagen
valor_output <- 100  # Precio por unidad producida ($)
costo_laboral <- 50  # Costo por hora de trabajo ($)

# Datos semanales
unidades_producidas <- c(1124, 1310, 1092, 981)
costo_laboral_total <- c(12735, 14842, 10603, 9526)
horas_laborales <- c(254.7, 296.84, 212.06, 190.52)
materiales <- c(21041, 24523, 20442, 18364)
overhead <- c(8992, 10480, 8736, 7848)

# Calculamos el valor total producido
valor_total <- unidades_producidas * valor_output

# Cálculo de insumo total para productividad multifactorial
insumo_total <- costo_laboral_total + materiales + overhead

# ---------------------------------------------
# Productividad Multifactorial (PMF)
# ---------------------------------------------
productividad_multifactorial <- valor_total / insumo_total

# ---------------------------------------------
# Productividad Laboral (PL)
# ---------------------------------------------
productividad_laboral <- unidades_producidas / horas_laborales

# ---------------------------------------------
# Resultados
# ---------------------------------------------
# Imprimir resultados intermedios
cat("\n--- Resultados de Productividad ---\n")
cat("Semana | Valor Total | Insumo Total | PMF  | Horas | PL  \n")
cat("--------------------------------------------------------\n")
for (i in 1:4) {
  cat(sprintf("  %d    | %10.2f | %12.2f | %.4f | %.2f | %.4f \n",
              i, valor_total[i], insumo_total[i], productividad_multifactorial[i], 
              horas_laborales[i], productividad_laboral[i]))
}

# Determinar la mejora en productividad
mejora_pmf <- ((productividad_multifactorial[4] - productividad_multifactorial[1]) / productividad_multifactorial[1]) * 100
mejora_pl <- ((productividad_laboral[4] - productividad_laboral[1]) / productividad_laboral[1]) * 100

cat("\n--- Análisis de Cambio en Productividad ---\n")
cat(sprintf("Cambio en Productividad Multifactorial: %.2f%%\n", mejora_pmf))
cat(sprintf("Cambio en Productividad Laboral: %.2f%%\n", mejora_pl))

# Conclusión
if (mejora_pmf > 0) {
  cat("\nLa productividad multifactorial ha mejorado a lo largo de las semanas.\n")
} else {
  cat("\nNo hubo mejora significativa en la productividad multifactorial.\n")
}

if (mejora_pl > 0) {
  cat("La productividad laboral también ha aumentado, indicando una mayor eficiencia del trabajo.\n")
} else {
  cat("No hubo mejora significativa en la productividad laboral.\n")
}
```

- **Producción con turnos extendidos**:

**Preguntas a resolver**

```r
# ------------------------------------------------------------------------------
# Cálculo de productividad multifactorial, productividad laboral y utilidades
# Administración de operaciones - Evaluación del impacto en productividad
# ------------------------------------------------------------------------------

# Definimos las variables según los datos del problema

# Producción original
unidades_original <- 2500
costo_produccion_original <- 120
precio_venta <- 200
empleados_turno1 <- 70
empleados_turno2 <- 30
horas_trabajadas_original <- 40 * (empleados_turno1 + empleados_turno2)
costo_total_original <- unidades_original * costo_produccion_original

# Producción en sobretiempo
unidades_sobrecarga <- 4000
costo_produccion_sobrecarga <- 144
horas_trabajadas_sobrecarga <- 72 * (empleados_turno1 + empleados_turno2)
costo_total_sobrecarga <- unidades_sobrecarga * costo_produccion_sobrecarga

# Cálculo de productividad multifactorial (PMF)
pmf_original <- unidades_original / costo_total_original
pmf_sobrecarga <- unidades_sobrecarga / costo_total_sobrecarga
cambio_pmf <- ((pmf_sobrecarga - pmf_original) / pmf_original) * 100

# Cálculo de productividad laboral (PL)
pl_original <- unidades_original / horas_trabajadas_original
pl_sobrecarga <- unidades_sobrecarga / horas_trabajadas_sobrecarga
cambio_pl <- ((pl_sobrecarga - pl_original) / pl_original) * 100

# Cálculo de utilidades semanales
utilidad_original <- (precio_venta - costo_produccion_original) * unidades_original
utilidad_sobrecarga <- (precio_venta - costo_produccion_sobrecarga) * unidades_sobrecarga
cambio_utilidad <- ((utilidad_sobrecarga - utilidad_original) / utilidad_original) * 100

# ------------------------------------------------------------------------------
# Mostramos los resultados
# ------------------------------------------------------------------------------

cat("**Resultados del análisis de productividad y utilidad**\n\n")

cat("1. Productividad Multifactorial (PMF):\n")
cat(sprintf("   - Productividad original: %.6f\n", pmf_original))
cat(sprintf("   - Productividad con sobrecarga: %.6f\n", pmf_sobrecarga))
cat(sprintf("   - Cambio en PMF: %.2f%%\n\n", cambio_pmf))

cat("2. Productividad Laboral (PL):\n")
cat(sprintf("   - Productividad original: %.6f\n", pl_original))
cat(sprintf("   - Productividad con sobrecarga: %.6f\n", pl_sobrecarga))
cat(sprintf("   - Cambio en PL: %.2f%%\n\n", cambio_pl))

cat("3. Utilidad Semanal:\n")
cat(sprintf("   - Utilidad original: $%.2f\n", utilidad_original))
cat(sprintf("   - Utilidad con sobrecarga: $%.2f\n", utilidad_sobrecarga))
cat(sprintf("   - Cambio en utilidad: %.2f%%\n\n", cambio_utilidad))

# ------------------------------------------------------------------------------
# Análisis de los resultados
# ------------------------------------------------------------------------------
cat("Conclusiones:\n")
if (cambio_pmf < 0) {
  cat("   - La productividad multifactorial disminuyó en", round(cambio_pmf, 2), "%.\n")
} else {
  cat("   - La productividad multifactorial aumentó en", round(cambio_pmf, 2), "%.\n")
}

if (cambio_pl < 0) {
  cat("   - La productividad laboral disminuyó en", round(cambio_pl, 2), "%.\n")
} else {
  cat("   - La productividad laboral aumentó en", round(cambio_pl, 2), "%.\n")
}

if (cambio_utilidad > 0) {
  cat("   - Las utilidades semanales aumentaron en", round(cambio_utilidad, 2), "%.\n")
} else {
  cat("   - Las utilidades semanales disminuyeron en", round(cambio_utilidad, 2), "%.\n")
}

cat("\nInterpretación: A pesar de la disminución en la eficiencia, el incremento en la producción permitió aumentar las utilidades semanales.\n")
```
