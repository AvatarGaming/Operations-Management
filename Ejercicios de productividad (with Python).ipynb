## ğŸ“ˆ Calcular y analizar la productividad multifactorial y la productividad laboral en un curso tÃ­pico de 3 crÃ©ditos en la Universidad Boehring.

- **ğŸ’µ MatrÃ­cula estudiantil:** $150 por hora crÃ©dito del semestre.
- **ğŸ›ï¸ Apoyo estatal:** $100 por hora crÃ©dito del semestre.
- **ğŸ‘¥ TamaÃ±o promedio de la clase:** 50 estudiantes.
- **ğŸ’¼ Costos laborales:** $4,000 por clase.
- **ğŸ“š Costos de materiales:** $20 por estudiante por clase.
- **ğŸ¢ Costos generales:** $25,000 por clase.

### ğŸ“ Preguntas:

a. Â¿CuÃ¡l es la relaciÃ³n de **productividad multifactorial** para este proceso de curso?  
b. Si los instructores trabajan un promedio de 14 horas por semana durante 16 semanas para cada clase de 3 crÃ©ditos con 50 estudiantes, Â¿cuÃ¡l es la relaciÃ³n de **productividad laboral**?

### ğŸ“Š Tabla 1: Datos del problema

| Variable                        | DescripciÃ³n                                      | Valor   |
|---------------------------------|--------------------------------------------------|---------|
| **Cuota escolar por hora crÃ©dito** | Costo que pagan los estudiantes por hora crÃ©dito | $150    |
| **Apoyo estatal por hora crÃ©dito** | Cantidad que aporta el gobierno estatal por hora crÃ©dito | $100    |
| **TamaÃ±o promedio del grupo**   | NÃºmero de estudiantes por grupo                  | 50      |
| **NÃºmero de crÃ©ditos por curso** | NÃºmero de crÃ©ditos por curso                     | 3       |

```python
# ğŸ“˜ Ejercicio 01 - Productividad en un Curso Universitario

# ğŸ“Š DefiniciÃ³n de variables
# Costos y beneficios del curso
cuota_escolar = 150  # ğŸ’° Costo por hora crÃ©dito pagado por estudiante (USD)
apoyo_estatal = 100  # ğŸ›ï¸ Subsidio estatal por hora crÃ©dito (USD)
tamano_grupo = 50    # ğŸ‘¨â€ğŸ“ NÃºmero de estudiantes por clase
creditos_curso = 3   # ğŸ“š NÃºmero de crÃ©ditos por curso

# ğŸ·ï¸ Costos asociados
costo_mano_obra = 4000    # ğŸ§‘â€ğŸ« Costo laboral por clase (USD)
costo_materiales = 20     # ğŸ“– Costo de materiales por estudiante (USD)
costos_generales = 25000  # ğŸ¢ Costos generales por clase (USD)

# â³ Horas trabajadas por el instructor
total_semanas = 16          # ğŸ“… DuraciÃ³n del curso en semanas
horas_por_semana = 14       # â±ï¸ Horas de trabajo del instructor por semana
horas_trabajadas = horas_por_semana * total_semanas  # â³ Total de horas trabajadas

# ğŸ’¡ CÃ¡lculo del valor del producto (Outputs)
total_cuota_escolar = cuota_escolar * creditos_curso * tamano_grupo  # Ingresos por cuota escolar
apoyo_estatal_total = apoyo_estatal * creditos_curso * tamano_grupo  # Ingresos por subsidio estatal
valor_producto_total = total_cuota_escolar + apoyo_estatal_total  # ğŸ’µ Ingresos totales del curso

# ğŸ’¡ CÃ¡lculo del valor de los insumos (Inputs)
costo_materiales_total = costo_materiales * tamano_grupo  # ğŸ“– Costo total de materiales
valor_insumos_total = costo_mano_obra + costo_materiales_total + costos_generales  # ğŸ’° Costos totales

# ğŸ“ˆ CÃ¡lculo de la productividad multifactorial
# ğŸ”¹ Productividad multifactorial = Valor del Producto / Valor de los Insumos
productividad_multifactorial = valor_producto_total / valor_insumos_total

# ğŸ“ˆ CÃ¡lculo de la productividad laboral
# ğŸ”¹ Productividad laboral = Valor del Producto / Horas trabajadas
productividad_laboral = valor_producto_total / horas_trabajadas

# ğŸ“¢ ImpresiÃ³n de resultados
print("=== ğŸ“Š Resultados del Ejercicio ===\n")

print("ğŸ“Œ Valor del Producto (Outputs):")
print("ğŸ“ Cuota Escolar Total: $", total_cuota_escolar)
print("ğŸ›ï¸ Apoyo Estatal Total: $", apoyo_estatal_total)
print("ğŸ’µ Valor Total del Producto: $", valor_producto_total, "\n")

print("ğŸ“Œ Valor de los Insumos (Inputs):")
print("ğŸ§‘â€ğŸ« Costo de Mano de Obra: $", costo_mano_obra)
print("ğŸ“– Costo de Materiales Total: $", costo_materiales_total)
print("ğŸ¢ Costos Generales: $", costos_generales)
print("ğŸ’° Valor Total de los Insumos: $", valor_insumos_total, "\n")

print("ğŸ“Œ Productividad Multifactorial:")
print("âš–ï¸ Productividad = Valor del Producto / Valor de los Insumos =", productividad_multifactorial, "\n")

print("ğŸ“Œ Productividad Laboral:")
print("â³ Horas trabajadas por instructor: ", horas_trabajadas, " horas")
print("ğŸ’µ Ingresos por curso: $", valor_producto_total)
print("ğŸ“Š Productividad Laboral (USD por hora):", productividad_laboral)
```

- **Productividad Laboral** = **Valor Total de la ProducciÃ³n** / **Horas Trabajadas**
  
  Donde:
  - **Valor Total de la ProducciÃ³n** = (NÃºmero de prendas de segunda Ã— Precio de venta de segunda) + (NÃºmero de prendas restantes Ã— Precio de venta restante)

---

Calcular la **productividad laboral** en tÃ©rminos de **USD por hora** para el proceso de fabricaciÃ³n de Natalie Attire.

# ğŸ“Š Ejercicio 04: AnÃ¡lisis de Productividad Laboral en Suds and Duds Laundry


## ğŸ§® Objetivo
Analizar la productividad laboral en el servicio de lavanderÃ­a Suds and Duds Laundry.

## ğŸ“‹ Datos
Suds and Duds Laundry lavÃ³ y planchÃ³ las siguientes cantidades de camisas por semana:

| Semana | Trabajadores | Total de horas | Camisas |
|:------:|-------------:|--------------:|--------:|
| 1      | Sud y Dud    | 24            | 68      |
| 2      | Sud y Jud    | 46            | 130     |
| 3      | Sud, Dud y Jud | 62          | 152     |
| 4      | Sud, Dud y Jud | 51          | 125     |
| 5      | Dud y Jud    | 45            | 131     |

## ğŸ“ Tareas
a. Calcular la ratio de productividad laboral para cada semana.
b. Explicar el patrÃ³n de productividad laboral que muestran los datos.

## ğŸ’¡ Instrucciones adicionales
- Comprender, interpretar y organizar la informaciÃ³n
- Presentar los cÃ¡lculos de forma clara
- Utilizar formato adecuado para fÃ³rmulas matemÃ¡ticas
- Optimizar la presentaciÃ³n para GitHub

```python
# ğŸ­ **CÃ¡lculo de la Productividad Laboral - Suds and Duds Laundry**
# -----------------------------------------------------------
# ğŸ“Œ La productividad laboral se calcula como:
#    Productividad = ProducciÃ³n (output) / Horas de trabajo (input)

# ğŸ“Š Definir los datos del problema
semanas = [1, 2, 3, 4, 5]  # ğŸ“… NÃºmero de semanas
trabajadores = [2, 2, 3, 3, 2]  # ğŸ‘¥ NÃºmero de trabajadores por semana
horas_trabajo = [24, 46, 62, 51, 45]  # â³ Total de horas trabajadas por semana
produccion = [68, 130, 152, 125, 131]  # ğŸ‘” Camisas producidas por semana

# ğŸ“ˆ Calcular la productividad laboral (camisas por hora trabajada)
productividad = [p / h for p, h in zip(produccion, horas_trabajo)]

# ğŸ–¥ï¸ Imprimir los resultados de productividad laboral
print("\nğŸ“Š Productividad Laboral (camisas por hora):")
for semana, prod in zip(semanas, productividad):
    print(f"ğŸ“… Semana {semana}: {prod:.4f} camisas/hora")

# ğŸ”„ Analizar los cambios en productividad entre semanas
cambio_productividad = [None] + [((productividad[i] - productividad[i-1]) / productividad[i-1]) * 100 
                                  for i in range(1, len(productividad))]

# ğŸ“‰ Imprimir los cambios porcentuales de productividad
print("\nğŸ“‰ Cambio porcentual en productividad entre semanas:")
for i in range(1, len(semanas)):
    print(f"ğŸ“… Semana {semanas[i]} vs Semana {semanas[i-1]}: {cambio_productividad[i]:.2f}%")

# ğŸ§ AnÃ¡lisis de productividad laboral
print("\nğŸ” AnÃ¡lisis de la Productividad Laboral:")
print("- La productividad laboral disminuye en la Semana 3 cuando hay 3 trabajadores.")
print("- Esto sugiere que el tercer trabajador no contribuye significativamente al rendimiento.")
print("- PodrÃ­a deberse a restricciones de espacio, equipos o flujo de trabajo ineficiente.")
```

```python
# ğŸ“Œ **Ejercicio 05: OptimizaciÃ³n de Costos en White Tiger Electronics**
# =========================================================
# ğŸ”§ CÃ¡lculo de reducciÃ³n de costos para mejorar productividad multifactorial
# ğŸ“Š AdministraciÃ³n de operaciones - AnÃ¡lisis de eficiencia en costos
# =========================================================

# ğŸ“Œ DefiniciÃ³n de variables clave
precio_venta = 300       # ğŸ’° Precio de venta por unidad ($)
costo_total = 150        # ğŸ·ï¸ Costo total de producciÃ³n por unidad ($)
costo_mano_obra = 30     # ğŸ‘· Costo de mano de obra por unidad ($)
costo_materiales = 70    # ğŸ› ï¸ Costo de materiales por unidad ($)
costo_overhead = 50      # ğŸ­ Costos generales (Overhead) por unidad ($)
mejora_productividad = 1.1  # ğŸ“ˆ Mejora esperada del 10% en productividad multifactorial

# ğŸ“Š CÃ¡lculo de la nueva entrada de recursos requerida
nuevo_input = costo_total / mejora_productividad

# ğŸ“‰ CÃ¡lculo de la reducciÃ³n total necesaria en costos
reduccion_necesaria = costo_total - nuevo_input

# ğŸ”¢ CÃ¡lculo de reducciÃ³n en Materiales
reduccion_materiales = (reduccion_necesaria / costo_materiales) * 100
nuevo_costo_materiales = costo_materiales - reduccion_necesaria

# ğŸ”¢ CÃ¡lculo de reducciÃ³n en Mano de Obra
reduccion_mano_obra = (reduccion_necesaria / costo_mano_obra) * 100
nuevo_costo_mano_obra = costo_mano_obra - reduccion_necesaria

# ğŸ”¢ CÃ¡lculo de reducciÃ³n en Costos Generales (Overhead)
reduccion_overhead = (reduccion_necesaria / costo_overhead) * 100
nuevo_costo_overhead = costo_overhead - reduccion_necesaria

# ğŸ“‹ ImpresiÃ³n de resultados
print("=======================================")
print(" White Tiger Electronics - Resultados ")
print("=======================================\n")

print("ğŸ·ï¸ Costo total original de producciÃ³n por unidad:", costo_total)
print("ğŸ“Š Nuevo costo total necesario para lograr mejora del 10%:", round(nuevo_input, 4))
print("ğŸ“‰ ReducciÃ³n total requerida en costos:", round(reduccion_necesaria, 4), "\n")

print("ğŸ› ï¸ ReducciÃ³n necesaria en Materiales:")
print(" - Nuevo costo de materiales:", round(nuevo_costo_materiales, 4))
print(" - Porcentaje de reducciÃ³n:", round(reduccion_materiales, 2), "%\n")

print("ğŸ‘· ReducciÃ³n necesaria en Mano de Obra:")
print(" - Nuevo costo de mano de obra:", round(nuevo_costo_mano_obra, 4))
print(" - Porcentaje de reducciÃ³n:", round(reduccion_mano_obra, 2), "%\n")

print("ğŸ­ ReducciÃ³n necesaria en Costos Generales (Overhead):")
print(" - Nuevo costo de overhead:", round(nuevo_costo_overhead, 4))
print(" - Porcentaje de reducciÃ³n:", round(reduccion_overhead, 2), "%")
```

# ğŸ­ Ejercicio 06: AnÃ¡lisis de Productividad en Symtecks


## ğŸ“Š Problema
Analizar el impacto de las mejoras recientes en un proceso especÃ­fico de Symtecks, utilizando la productividad multifactorial y la productividad laboral como medidas de desempeÃ±o.

## ğŸ“‹ Contexto
En Symtecks, el resultado de un proceso especÃ­fico estÃ¡ valorado en $100 por unidad. El costo de mano de obra es $50 por hora, incluyendo beneficios. El departamento de contabilidad proporcionÃ³ la siguiente informaciÃ³n sobre el proceso durante las Ãºltimas cuatro semanas:

| Concepto | Semana 1 | Semana 2 | Semana 3 | Semana 4 |
|:-----------------|----------:|----------:|----------:|----------:|
| Unidades Producidas | 1,124 | 1,310 | 1,092 | 981 |
| Mano de Obra ($) | 12,735 | 14,842 | 10,603 | 9,526 |
| Materiales ($) | 21,041 | 24,523 | 20,442 | 18,364 |
| Gastos Generales ($) | 8,992 | 10,480 | 8,736 | 7,848 |

## ğŸ“ Tareas

a. Utilizar la ratio de productividad multifactorial para determinar si las mejoras recientes del proceso tuvieron algÃºn efecto y, en caso afirmativo, cuÃ¡ndo fue notable dicho efecto.

b. Â¿Ha cambiado la productividad laboral? Utilizar la ratio de productividad laboral para respaldar su respuesta.

## ğŸ’¡ Instrucciones
- Calcular los indicadores de productividad para cada semana
- Analizar las tendencias observadas
- Interpretar los resultados en el contexto de mejoras de proceso
- Presentar conclusiones claras y fundamentadas

```python
# ğŸ­ **CÃ¡lculo de Productividad Multifactorial y Laboral**
# -------------------------------------------------------
# ğŸ“Œ Evaluamos la eficiencia del proceso en Symtecks mediante dos mÃ©tricas clave:
#    1ï¸âƒ£ Productividad Multifactorial (PMF): RelaciÃ³n entre el valor de producciÃ³n y el costo total de insumos.
#    2ï¸âƒ£ Productividad Laboral (PL): RelaciÃ³n entre unidades producidas y horas trabajadas.

# ğŸ“Š **Datos Iniciales**
valor_output = 100  # Precio por unidad producida ($)
costo_laboral = 50   # Costo por hora de trabajo ($)

# ğŸ”¹ **Datos Semanales**
unidades_producidas = [1124, 1310, 1092, 981]
costo_laboral_total = [12735, 14842, 10603, 9526]
horas_laborales = [254.7, 296.84, 212.06, 190.52]
materiales = [21041, 24523, 20442, 18364]
overhead = [8992, 10480, 8736, 7848]

# ğŸ” **CÃ¡lculos**
# -------------------------------------------------------
# ğŸ“Œ **Paso 1: Calcular el valor total de producciÃ³n**
valor_total = [u * valor_output for u in unidades_producidas]

# ğŸ“Œ **Paso 2: Calcular el insumo total** (suma de costos laborales, materiales y overhead)
insumo_total = [cl + m + o for cl, m, o in zip(costo_laboral_total, materiales, overhead)]

# ğŸ“Œ **Paso 3: Calcular la Productividad Multifactorial (PMF)**
productividad_multifactorial = [vt / it for vt, it in zip(valor_total, insumo_total)]

# ğŸ“Œ **Paso 4: Calcular la Productividad Laboral (PL)**
productividad_laboral = [u / h for u, h in zip(unidades_producidas, horas_laborales)]

# ğŸ“Š **Resultados**
# -------------------------------------------------------
print("\n--- ğŸ“ˆ Resultados de Productividad ---")
print("Semana | Valor Total | Insumo Total | PMF    | Horas  | PL    ")
print("----------------------------------------------------------")
for i in range(4):
    print(f"  {i+1}    | {valor_total[i]:10.2f} | {insumo_total[i]:12.2f} | {productividad_multifactorial[i]:.4f} | {horas_laborales[i]:.2f} | {productividad_laboral[i]:.4f}")

# ğŸ” **AnÃ¡lisis de Mejora**
# -------------------------------------------------------
mejora_pmf = ((productividad_multifactorial[3] - productividad_multifactorial[0]) / productividad_multifactorial[0]) * 100
mejora_pl = ((productividad_laboral[3] - productividad_laboral[0]) / productividad_laboral[0]) * 100

print("\n--- ğŸ“Š AnÃ¡lisis de Cambio en Productividad ---")
print(f"ğŸ“Œ Cambio en Productividad Multifactorial: {mejora_pmf:.2f}%")
print(f"ğŸ“Œ Cambio en Productividad Laboral: {mejora_pl:.2f}%")

# ğŸ **ConclusiÃ³n**
# -------------------------------------------------------
if mejora_pmf > 0:
    print("\nâœ… La productividad multifactorial ha mejorado a lo largo de las semanas.")
else:
    print("\nâŒ No hubo mejora significativa en la productividad multifactorial.")

if mejora_pl > 0:
    print("âœ… La productividad laboral tambiÃ©n ha aumentado, indicando una mayor eficiencia del trabajo.")
else:
    print("âŒ No hubo mejora significativa en la productividad laboral.")
```

# ğŸ¢ Ejercicio 07: AnÃ¡lisis de Productividad en Big Black Bird Company



## ğŸ“ˆ Objetivo
Analizar el impacto de un cambio en el programa de producciÃ³n de la empresa Big Black Bird Company (BBBC) en tÃ©rminos de productividad multifactorial, productividad laboral y utilidades semanales.

## ğŸ“‹ Contexto
The Big Black Bird Company (BBBC) ha recibido un gran pedido de uniformes militares especiales con revestimiento plÃ¡stico para ser utilizados en una operaciÃ³n militar urgente.

### SituaciÃ³n Inicial
| Concepto | Valor |
|:----------------------------------------|------:|
| ProducciÃ³n normal semanal | 2,500 uniformes |
| Costo estÃ¡ndar por unidad | $120 |
| Precio del contrato por uniforme | $200 |
| Empleados primer turno | 70 |
| Empleados segundo turno | 30 |
| Horas de trabajo por turno | 40 horas/semana |

### SituaciÃ³n Modificada (ProducciÃ³n 24/7)
| Concepto | Valor |
|:----------------------------------------|------:|
| ProducciÃ³n con turnos extendidos | 4,000 uniformes |
| Costo por unidad con turnos extendidos | $144 |
| DÃ­as de trabajo | 6 dÃ­as/semana |
| Horas por turno | 72 horas/semana |

## ğŸ“ Tareas
a. Â¿La ratio de productividad multifactorial aumentÃ³, disminuyÃ³ o permaneciÃ³ igual? Si cambiÃ³, Â¿en quÃ© porcentaje cambiÃ³?

b. Â¿La ratio de productividad laboral aumentÃ³, disminuyÃ³ o permaneciÃ³ igual? Si cambiÃ³, Â¿en quÃ© porcentaje cambiÃ³?

c. Â¿Las utilidades semanales aumentaron, disminuyeron o permanecieron iguales?

## ğŸ’¡ Instrucciones adicionales
- Calcular los indicadores para ambos escenarios
- Comparar los resultados y determinar los porcentajes de cambio
- Analizar el impacto econÃ³mico de la decisiÃ³n
- Presentar conclusiones claras y fundamentadas

```python
# ğŸ­ **AnÃ¡lisis de Productividad y Utilidades en BBBC**
# -----------------------------------------------------
# ğŸ“Œ Evaluamos el impacto del cambio en el programa de producciÃ³n de BBBC
#    en tÃ©rminos de productividad multifactorial, productividad laboral y utilidades semanales.

# -----------------------------------------------------
# ğŸ“Š Definimos las variables segÃºn los datos del problema
# -----------------------------------------------------

# ğŸ”¹ ProducciÃ³n Normal
unidades_original = 2500  # Uniformes producidos por semana
costo_produccion_original = 120  # Costo por uniforme ($)
precio_venta = 200  # Precio de venta por uniforme ($)

# ğŸ‘· DistribuciÃ³n de empleados y horas trabajadas
empleados_turno1 = 70  # Empleados en el turno 1
empleados_turno2 = 30  # Empleados en el turno 2
horas_trabajadas_original = 40 * (empleados_turno1 + empleados_turno2)  # Horas totales trabajadas

# ğŸ’° Costos Totales
costo_total_original = unidades_original * costo_produccion_original  # Costo total de producciÃ³n ($)

# ğŸ”º ProducciÃ³n con Turnos Extendidos
unidades_sobrecarga = 4000  # Uniformes producidos por semana con turnos extendidos
costo_produccion_sobrecarga = 144  # Nuevo costo por uniforme ($)
horas_trabajadas_sobrecarga = 72 * (empleados_turno1 + empleados_turno2)  # Horas totales trabajadas en sobrecarga
costo_total_sobrecarga = unidades_sobrecarga * costo_produccion_sobrecarga  # Costo total de producciÃ³n ($)

# -----------------------------------------------------
# ğŸ“Š CÃ¡lculo de Productividad Multifactorial (PMF)
# -----------------------------------------------------
pmf_original = unidades_original / costo_total_original  # Productividad original
pmf_sobrecarga = unidades_sobrecarga / costo_total_sobrecarga  # Productividad con turnos extendidos
cambio_pmf = ((pmf_sobrecarga - pmf_original) / pmf_original) * 100  # Cambio porcentual en PMF

# -----------------------------------------------------
# ğŸ‘¨â€ğŸ­ CÃ¡lculo de Productividad Laboral (PL)
# -----------------------------------------------------
pl_original = unidades_original / horas_trabajadas_original  # Productividad laboral original
pl_sobrecarga = unidades_sobrecarga / horas_trabajadas_sobrecarga  # Productividad laboral con sobrecarga
cambio_pl = ((pl_sobrecarga - pl_original) / pl_original) * 100  # Cambio porcentual en PL

# -----------------------------------------------------
# ğŸ’° CÃ¡lculo de Utilidades Semanales
# -----------------------------------------------------
utilidad_original = (precio_venta - costo_produccion_original) * unidades_original  # Utilidad semanal original ($)
utilidad_sobrecarga = (precio_venta - costo_produccion_sobrecarga) * unidades_sobrecarga  # Utilidad semanal con sobrecarga ($)
cambio_utilidad = ((utilidad_sobrecarga - utilidad_original) / utilidad_original) * 100  # Cambio porcentual en utilidad

# -----------------------------------------------------
# ğŸ“¢ Mostramos los resultados
# -----------------------------------------------------
print("ğŸ­ **Resultados del AnÃ¡lisis de Productividad y Utilidad**\n")

print("ğŸ“Š 1. Productividad Multifactorial (PMF):")
print(f"   - Productividad original: {pmf_original:.6f}")
print(f"   - Productividad con sobrecarga: {pmf_sobrecarga:.6f}")
print(f"   - Cambio en PMF: {cambio_pmf:.2f}%\n")

print("ğŸ‘¨â€ğŸ­ 2. Productividad Laboral (PL):")
print(f"   - Productividad original: {pl_original:.6f}")
print(f"   - Productividad con sobrecarga: {pl_sobrecarga:.6f}")
print(f"   - Cambio en PL: {cambio_pl:.2f}%\n")

print("ğŸ’° 3. Utilidad Semanal:")
print(f"   - Utilidad original: ${utilidad_original:.2f}")
print(f"   - Utilidad con sobrecarga: ${utilidad_sobrecarga:.2f}")
print(f"   - Cambio en utilidad: {cambio_utilidad:.2f}%\n")

# -----------------------------------------------------
# ğŸ” AnÃ¡lisis de los Resultados
# -----------------------------------------------------
print("ğŸ“Œ **Conclusiones:**")
if cambio_pmf < 0:
    print(f"   - La productividad multifactorial disminuyÃ³ en {round(cambio_pmf, 2)}% ğŸš¨.")
else:
    print(f"   - La productividad multifactorial aumentÃ³ en {round(cambio_pmf, 2)}% âœ….")

if cambio_pl < 0:
    print(f"   - La productividad laboral disminuyÃ³ en {round(cambio_pl, 2)}% ğŸš¨.")
else:
    print(f"   - La productividad laboral aumentÃ³ en {round(cambio_pl, 2)}% âœ….")

if cambio_utilidad > 0:
    print(f"   - Las utilidades semanales aumentaron en {round(cambio_utilidad, 2)}% âœ….")
else:
    print(f"   - Las utilidades semanales disminuyeron en {round(cambio_utilidad, 2)}% ğŸš¨.")

print("\nğŸ“Š **InterpretaciÃ³n:** A pesar de la reducciÃ³n en la eficiencia, el aumento en la producciÃ³n permitiÃ³ incrementar las utilidades semanales.")
```
