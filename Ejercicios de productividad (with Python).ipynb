# 📊 Ejercicio 01

---

## 📈 Calcular y analizar la productividad multifactorial y la productividad laboral en un curso típico de 3 créditos en la Universidad Boehring.

- **💵 Matrícula estudiantil:** $150 por hora crédito del semestre.
- **🏛️ Apoyo estatal:** $100 por hora crédito del semestre.
- **👥 Tamaño promedio de la clase:** 50 estudiantes.
- **💼 Costos laborales:** $4,000 por clase.
- **📚 Costos de materiales:** $20 por estudiante por clase.
- **🏢 Costos generales:** $25,000 por clase.

### 📝 Preguntas:

a. ¿Cuál es la relación de **productividad multifactorial** para este proceso de curso?  
b. Si los instructores trabajan un promedio de 14 horas por semana durante 16 semanas para cada clase de 3 créditos con 50 estudiantes, ¿cuál es la relación de **productividad laboral**?

### 📊 Tabla 1: Datos del problema

| Variable                        | Descripción                                      | Valor   |
|---------------------------------|--------------------------------------------------|---------|
| **Cuota escolar por hora crédito** | Costo que pagan los estudiantes por hora crédito | $150    |
| **Apoyo estatal por hora crédito** | Cantidad que aporta el gobierno estatal por hora crédito | $100    |
| **Tamaño promedio del grupo**   | Número de estudiantes por grupo                  | 50      |
| **Número de créditos por curso** | Número de créditos por curso                     | 3       |

```python
# 📘 Ejercicio 01 - Productividad en un Curso Universitario

# 📊 Definición de variables
# Costos y beneficios del curso
cuota_escolar = 150  # 💰 Costo por hora crédito pagado por estudiante (USD)
apoyo_estatal = 100  # 🏛️ Subsidio estatal por hora crédito (USD)
tamano_grupo = 50    # 👨‍🎓 Número de estudiantes por clase
creditos_curso = 3   # 📚 Número de créditos por curso

# 🏷️ Costos asociados
costo_mano_obra = 4000    # 🧑‍🏫 Costo laboral por clase (USD)
costo_materiales = 20     # 📖 Costo de materiales por estudiante (USD)
costos_generales = 25000  # 🏢 Costos generales por clase (USD)

# ⏳ Horas trabajadas por el instructor
total_semanas = 16          # 📅 Duración del curso en semanas
horas_por_semana = 14       # ⏱️ Horas de trabajo del instructor por semana
horas_trabajadas = horas_por_semana * total_semanas  # ⏳ Total de horas trabajadas

# 💡 Cálculo del valor del producto (Outputs)
total_cuota_escolar = cuota_escolar * creditos_curso * tamano_grupo  # Ingresos por cuota escolar
apoyo_estatal_total = apoyo_estatal * creditos_curso * tamano_grupo  # Ingresos por subsidio estatal
valor_producto_total = total_cuota_escolar + apoyo_estatal_total  # 💵 Ingresos totales del curso

# 💡 Cálculo del valor de los insumos (Inputs)
costo_materiales_total = costo_materiales * tamano_grupo  # 📖 Costo total de materiales
valor_insumos_total = costo_mano_obra + costo_materiales_total + costos_generales  # 💰 Costos totales

# 📈 Cálculo de la productividad multifactorial
# 🔹 Productividad multifactorial = Valor del Producto / Valor de los Insumos
productividad_multifactorial = valor_producto_total / valor_insumos_total

# 📈 Cálculo de la productividad laboral
# 🔹 Productividad laboral = Valor del Producto / Horas trabajadas
productividad_laboral = valor_producto_total / horas_trabajadas

# 📢 Impresión de resultados
print("=== 📊 Resultados del Ejercicio ===\n")

print("📌 Valor del Producto (Outputs):")
print("🎓 Cuota Escolar Total: $", total_cuota_escolar)
print("🏛️ Apoyo Estatal Total: $", apoyo_estatal_total)
print("💵 Valor Total del Producto: $", valor_producto_total, "\n")

print("📌 Valor de los Insumos (Inputs):")
print("🧑‍🏫 Costo de Mano de Obra: $", costo_mano_obra)
print("📖 Costo de Materiales Total: $", costo_materiales_total)
print("🏢 Costos Generales: $", costos_generales)
print("💰 Valor Total de los Insumos: $", valor_insumos_total, "\n")

print("📌 Productividad Multifactorial:")
print("⚖️ Productividad = Valor del Producto / Valor de los Insumos =", productividad_multifactorial, "\n")

print("📌 Productividad Laboral:")
print("⏳ Horas trabajadas por instructor: ", horas_trabajadas, " horas")
print("💵 Ingresos por curso: $", valor_producto_total)
print("📊 Productividad Laboral (USD por hora):", productividad_laboral)
```

## 🧵 Calcular y analizar la productividad laboral en la producción de prendas de vestir en Natalie Attire.

Natalie Attire produce prendas de vestir de moda. Las prendas de segunda se venden a **$90** cada una en la tienda de fábrica de Attire. Las **80** prendas restantes se venden a distribuidores minoristas a **$200** cada una.  

### 📝 Pregunta:

¿Cuál es la relación de **productividad laboral** de este proceso de fabricación?

### 📊 Tabla 1: Datos del problema

| Variable                        | Valor       |
|---------------------------------|-------------|
| **Número total de prendas**     | 132         |
| **Número de prendas de segunda**| 52          |

| **Precio de venta por prenda de segunda** | $90  |
| **Precio de venta por prenda restante**  | $200 |
| **Horas de trabajo empleadas**  | 360         |

---

### 📌 Fórmulas clave:

- **Productividad Laboral** = **Valor Total de la Producción** / **Horas Trabajadas**
  
  Donde:
  - **Valor Total de la Producción** = (Número de prendas de segunda × Precio de venta de segunda) + (Número de prendas restantes × Precio de venta restante)

---

### 🎯 Objetivo:

Calcular la **productividad laboral** en términos de **USD por hora** para el proceso de fabricación de Natalie Attire.


```python
# 👗 **Ejercicio 02: Productividad Laboral en la Industria Textil**

# 📌 Definir variables clave del problema
# Datos de entrada sobre la producción y costos en Natalie Attire
num_prendas_totales = 132  # Total de prendas producidas 👕
num_prendas_segunda = 52   # Prendas defectuosas (segunda calidad) ❌
num_prendas_buenas = 80    # Prendas sin defectos ✅
precio_segunda = 90        # Precio de venta por prenda de segunda 🏷️
precio_buenas = 200        # Precio de venta por prenda sin defectos 🛍️
horas_trabajo = 360        # Total de horas trabajadas ⏳

# 💰 Cálculo del valor total de la producción
# Se obtiene multiplicando la cantidad de cada tipo de prenda por su precio de venta
valor_prendas_segunda = num_prendas_segunda * precio_segunda  # Valor total de prendas de segunda
valor_prendas_buenas = num_prendas_buenas * precio_buenas     # Valor total de prendas sin defectos

# 🔢 Sumar ambos valores para obtener el ingreso total generado por la producción
valor_total_producto = valor_prendas_segunda + valor_prendas_buenas

# 📊 Cálculo de la productividad laboral
# Fórmula: Productividad laboral = Valor total de la producción / Horas trabajadas
productividad_laboral = valor_total_producto / horas_trabajo

# 📢 Mostrar los resultados en consola con formato claro
print("---- 📊 Resultados ----\n")
print("🧾 Valor total de prendas de segunda: $", valor_prendas_segunda)
print("🧾 Valor total de prendas buenas: $", valor_prendas_buenas)
print("💰 Valor total del producto: $", valor_total_producto)
print("⏳ Horas de trabajo empleadas:", horas_trabajo, "horas")
print("📈 Productividad laboral: $", round(productividad_laboral, 2), " por hora")
```

# 📊 Ejercicio 03 - Análisis de Productividad en la Universidad Boehring

## 🏈 Tendencias y Desafíos en la Gestión de Operaciones

Coach Bjourn Toulouse ha llevado a los Big Red Herrings a varias temporadas decepcionantes de fútbol. Solo una mejor contratación devolverá a los Big Red Herrings a su forma ganadora. Debido al estado actual del programa, es poco probable que los fanáticos de la Universidad Boehring apoyen aumentos en el precio de las entradas de temporada, que actualmente es de **$192**.  

La mejora en la contratación aumentará los costos generales a **$30,000** por sección de clase, desde los **$25,000** actuales. El plan presupuestario de la universidad es cubrir los costos de contratación aumentando el tamaño promedio de la clase a **75 estudiantes**.  

- **Costos laborales:** Aumentarán a **$6,500** por curso de 3 créditos.  
- **Costos de materiales:** Serán de **$25** por estudiante para cada curso de 3 créditos.  
- **Matrícula:** Será de **$200** por crédito semestral, complementada con un apoyo estatal de **$100** por crédito semestral.  

### 📝 Preguntas:

a. ¿Cuál es la relación de **productividad multifactorial**? En comparación con el resultado obtenido en el Problema Resuelto 1, ¿la productividad aumentó o disminuyó para el proceso del curso?  
b. Si los instructores trabajan un promedio de **20 horas por semana** durante **16 semanas** para cada clase de 3 créditos con **75 estudiantes**, ¿cuál es la relación de **productividad laboral**?

---

### 📊 Datos del Problema

| Variable                        | Valor       |
|---------------------------------|-------------|
| **Precio de la entrada de temporada** | $192       |
| **Costos generales por sección** | $30,000     |
| **Tamaño promedio de la clase**  | 75 estudiantes |
| **Costos laborales por curso**   | $6,500      |
| **Costos de materiales por estudiante** | $25   |
| **Matrícula por crédito**        | $200        |
| **Apoyo estatal por crédito**    | $100        |
| **Horas de trabajo por semana**  | 20          |
| **Duración del curso (semanas)** | 16          |

---

### 📌 Fórmulas clave:

- **Productividad Multifactorial** = **Valor del Producto** / **Valor de los Insumos**  
  Donde:  
  - **Valor del Producto** = (Matrícula + Apoyo estatal) × Número de créditos × Número de estudiantes  
  - **Valor de los Insumos** = Costos laborales + (Costos de materiales × Número de estudiantes) + Costos generales  

- **Productividad Laboral** = **Valor del Producto** / **Horas Trabajadas**  
  Donde:  
  - **Horas Trabajadas** = Horas por semana × Semanas del curso  

---

### 🎯 Objetivo:

Calcular la **productividad multifactorial** y la **productividad laboral** en el nuevo escenario de la Universidad Boehring, y comparar los resultados con el Problema Resuelto 1.


```python
# ========================================================
# 🎓 Análisis de Productividad Multifactorial y Laboral en Python
# ========================================================

# 📌 Administración de Operaciones: Evaluamos cómo los cambios 
# en costos y horas de trabajo afectan la productividad.

# ----------------------------------
# 📊 Definición de Variables Iniciales
# ----------------------------------

# 🔢 Costos y parámetros originales
cuota_escolar_original = 150     # 💰 Costo de matrícula por crédito
apoyo_estatal_original = 100     # 🏛️ Apoyo estatal por crédito
tamano_grupo_original = 50       # 🎓 Tamaño de la clase
creditos_curso = 3               # 📚 Créditos por curso
costo_mano_obra_original = 4000  # 💼 Costos laborales por curso
costo_generales_original = 25000 # 🏛️ Costos generales por sección
costo_materiales_original = 1000 # 📖 Costos de materiales
horas_trabajo_original = 224     # ⏳ Horas de trabajo del instructor

# 🔢 Costos y parámetros ajustados tras los cambios
cuota_escolar_ajuste = 200       # 💰 Nueva matrícula por crédito
apoyo_estatal_ajuste = 100       # 🏛️ Apoyo estatal sin cambios
tamano_grupo_ajuste = 75         # 🎓 Nuevo tamaño de clase
costo_mano_obra_ajuste = 6500    # 💼 Nuevos costos laborales
costo_generales_ajuste = 30000   # 🏛️ Nuevos costos generales
costo_materiales_ajuste = 1875   # 📖 Nuevos costos de materiales
horas_trabajo_ajuste = 320       # ⏳ Nuevas horas de trabajo

# ----------------------------------
# 📈 Cálculo del Valor del Producto (Ingresos Generados)
# ----------------------------------
valor_producto_original = (cuota_escolar_original + apoyo_estatal_original) * creditos_curso * tamano_grupo_original
valor_producto_ajuste = (cuota_escolar_ajuste + apoyo_estatal_ajuste) * creditos_curso * tamano_grupo_ajuste

# ----------------------------------
# 💵 Cálculo de los Insumos (Costos Totales)
# ----------------------------------
insumos_original = costo_mano_obra_original + costo_generales_original + costo_materiales_original
insumos_ajuste = costo_mano_obra_ajuste + costo_generales_ajuste + costo_materiales_ajuste

# ----------------------------------
# 📊 Cálculo del Índice de Productividad Multifactorial
# ----------------------------------
productividad_multifactorial_original = valor_producto_original / insumos_original
productividad_multifactorial_ajuste = valor_producto_ajuste / insumos_ajuste

# ----------------------------------
# ⏳ Cálculo de la Productividad Laboral (por Hora Trabajada)
# ----------------------------------
productividad_laboral_original = valor_producto_original / horas_trabajo_original
productividad_laboral_ajuste = valor_producto_ajuste / horas_trabajo_ajuste

# ----------------------------------
# 📉 Cálculo del Cambio Porcentual en la Productividad
# ----------------------------------
cambio_productividad_multifactorial = ((productividad_multifactorial_ajuste - productividad_multifactorial_original) / productividad_multifactorial_original) * 100
cambio_productividad_laboral = ((productividad_laboral_ajuste - productividad_laboral_original) / productividad_laboral_original) * 100

# ----------------------------------
# 📢 Resultados del Análisis
# ----------------------------------
print("\n--- 📊 Resultados de Productividad ---\n")
print("💰 Valor del Producto (Original):", valor_producto_original)
print("💰 Valor del Producto (Ajustado):", valor_producto_ajuste)
print("📊 Productividad Multifactorial Original:", round(productividad_multifactorial_original, 4))
print("📊 Productividad Multifactorial Ajustada:", round(productividad_multifactorial_ajuste, 4))
print("📉 Cambio en Productividad Multifactorial:", round(cambio_productividad_multifactorial, 2), "%")
print("⏳ Productividad Laboral Original:", round(productividad_laboral_original, 4))
print("⏳ Productividad Laboral Ajustada:", round(productividad_laboral_ajuste, 4))
print("📉 Cambio en Productividad Laboral:", round(cambio_productividad_laboral, 2), "%")

# ========================================================
# 📌 CONCLUSIÓN:
# - 📈 Se observa un incremento en la productividad laboral y multifactorial.
# - 📊 El aumento de costos fue compensado por un mayor tamaño de grupo y eficiencia.
# - 🔍 Es crucial evaluar continuamente estos indicadores para optimizar recursos.
# ========================================================
```

# 📊 Ejercicio 04: Análisis de Productividad Laboral en Suds and Duds Laundry


## 🧮 Objetivo
Analizar la productividad laboral en el servicio de lavandería Suds and Duds Laundry.

## 📋 Datos
Suds and Duds Laundry lavó y planchó las siguientes cantidades de camisas por semana:

| Semana | Trabajadores | Total de horas | Camisas |
|:------:|-------------:|--------------:|--------:|
| 1      | Sud y Dud    | 24            | 68      |
| 2      | Sud y Jud    | 46            | 130     |
| 3      | Sud, Dud y Jud | 62          | 152     |
| 4      | Sud, Dud y Jud | 51          | 125     |
| 5      | Dud y Jud    | 45            | 131     |

## 📝 Tareas
a. Calcular la ratio de productividad laboral para cada semana.
b. Explicar el patrón de productividad laboral que muestran los datos.

## 💡 Instrucciones adicionales
- Comprender, interpretar y organizar la información
- Presentar los cálculos de forma clara
- Utilizar formato adecuado para fórmulas matemáticas
- Optimizar la presentación para GitHub

```python
# 🏭 **Cálculo de la Productividad Laboral - Suds and Duds Laundry**
# -----------------------------------------------------------
# 📌 La productividad laboral se calcula como:
#    Productividad = Producción (output) / Horas de trabajo (input)

# 📊 Definir los datos del problema
semanas = [1, 2, 3, 4, 5]  # 📅 Número de semanas
trabajadores = [2, 2, 3, 3, 2]  # 👥 Número de trabajadores por semana
horas_trabajo = [24, 46, 62, 51, 45]  # ⏳ Total de horas trabajadas por semana
produccion = [68, 130, 152, 125, 131]  # 👔 Camisas producidas por semana

# 📈 Calcular la productividad laboral (camisas por hora trabajada)
productividad = [p / h for p, h in zip(produccion, horas_trabajo)]

# 🖥️ Imprimir los resultados de productividad laboral
print("\n📊 Productividad Laboral (camisas por hora):")
for semana, prod in zip(semanas, productividad):
    print(f"📅 Semana {semana}: {prod:.4f} camisas/hora")

# 🔄 Analizar los cambios en productividad entre semanas
cambio_productividad = [None] + [((productividad[i] - productividad[i-1]) / productividad[i-1]) * 100 
                                  for i in range(1, len(productividad))]

# 📉 Imprimir los cambios porcentuales de productividad
print("\n📉 Cambio porcentual en productividad entre semanas:")
for i in range(1, len(semanas)):
    print(f"📅 Semana {semanas[i]} vs Semana {semanas[i-1]}: {cambio_productividad[i]:.2f}%")

# 🧐 Análisis de productividad laboral
print("\n🔍 Análisis de la Productividad Laboral:")
print("- La productividad laboral disminuye en la Semana 3 cuando hay 3 trabajadores.")
print("- Esto sugiere que el tercer trabajador no contribuye significativamente al rendimiento.")
print("- Podría deberse a restricciones de espacio, equipos o flujo de trabajo ineficiente.")
```

```python
# 🏭 **Cálculo de Productividad Multifactorial y Laboral**
# -------------------------------------------------------
# 📌 Evaluamos la eficiencia del proceso en Symtecks mediante dos métricas clave:
#    1️⃣ Productividad Multifactorial (PMF): Relación entre el valor de producción y el costo total de insumos.
#    2️⃣ Productividad Laboral (PL): Relación entre unidades producidas y horas trabajadas.

# 📊 **Datos Iniciales**
valor_output = 100  # Precio por unidad producida ($)
costo_laboral = 50   # Costo por hora de trabajo ($)

# 🔹 **Datos Semanales**
unidades_producidas = [1124, 1310, 1092, 981]
costo_laboral_total = [12735, 14842, 10603, 9526]
horas_laborales = [254.7, 296.84, 212.06, 190.52]
materiales = [21041, 24523, 20442, 18364]
overhead = [8992, 10480, 8736, 7848]

# 🔍 **Cálculos**
# -------------------------------------------------------
# 📌 **Paso 1: Calcular el valor total de producción**
valor_total = [u * valor_output for u in unidades_producidas]

# 📌 **Paso 2: Calcular el insumo total** (suma de costos laborales, materiales y overhead)
insumo_total = [cl + m + o for cl, m, o in zip(costo_laboral_total, materiales, overhead)]

# 📌 **Paso 3: Calcular la Productividad Multifactorial (PMF)**
productividad_multifactorial = [vt / it for vt, it in zip(valor_total, insumo_total)]

# 📌 **Paso 4: Calcular la Productividad Laboral (PL)**
productividad_laboral = [u / h for u, h in zip(unidades_producidas, horas_laborales)]

# 📊 **Resultados**
# -------------------------------------------------------
print("\n--- 📈 Resultados de Productividad ---")
print("Semana | Valor Total | Insumo Total | PMF    | Horas  | PL    ")
print("----------------------------------------------------------")
for i in range(4):
    print(f"  {i+1}    | {valor_total[i]:10.2f} | {insumo_total[i]:12.2f} | {productividad_multifactorial[i]:.4f} | {horas_laborales[i]:.2f} | {productividad_laboral[i]:.4f}")

# 🔍 **Análisis de Mejora**
# -------------------------------------------------------
mejora_pmf = ((productividad_multifactorial[3] - productividad_multifactorial[0]) / productividad_multifactorial[0]) * 100
mejora_pl = ((productividad_laboral[3] - productividad_laboral[0]) / productividad_laboral[0]) * 100

print("\n--- 📊 Análisis de Cambio en Productividad ---")
print(f"📌 Cambio en Productividad Multifactorial: {mejora_pmf:.2f}%")
print(f"📌 Cambio en Productividad Laboral: {mejora_pl:.2f}%")

# 🏁 **Conclusión**
# -------------------------------------------------------
if mejora_pmf > 0:
    print("\n✅ La productividad multifactorial ha mejorado a lo largo de las semanas.")
else:
    print("\n❌ No hubo mejora significativa en la productividad multifactorial.")

if mejora_pl > 0:
    print("✅ La productividad laboral también ha aumentado, indicando una mayor eficiencia del trabajo.")
else:
    print("❌ No hubo mejora significativa en la productividad laboral.")
```

# 🏢 Ejercicio 07: Análisis de Productividad en Big Black Bird Company



## 📈 Objetivo
Analizar el impacto de un cambio en el programa de producción de la empresa Big Black Bird Company (BBBC) en términos de productividad multifactorial, productividad laboral y utilidades semanales.

## 📋 Contexto
The Big Black Bird Company (BBBC) ha recibido un gran pedido de uniformes militares especiales con revestimiento plástico para ser utilizados en una operación militar urgente.

### Situación Inicial
| Concepto | Valor |
|:----------------------------------------|------:|
| Producción normal semanal | 2,500 uniformes |
| Costo estándar por unidad | $120 |
| Precio del contrato por uniforme | $200 |
| Empleados primer turno | 70 |
| Empleados segundo turno | 30 |
| Horas de trabajo por turno | 40 horas/semana |

### Situación Modificada (Producción 24/7)
| Concepto | Valor |
|:----------------------------------------|------:|
| Producción con turnos extendidos | 4,000 uniformes |
| Costo por unidad con turnos extendidos | $144 |
| Días de trabajo | 6 días/semana |
| Horas por turno | 72 horas/semana |

## 📝 Tareas
a. ¿La ratio de productividad multifactorial aumentó, disminuyó o permaneció igual? Si cambió, ¿en qué porcentaje cambió?

b. ¿La ratio de productividad laboral aumentó, disminuyó o permaneció igual? Si cambió, ¿en qué porcentaje cambió?

c. ¿Las utilidades semanales aumentaron, disminuyeron o permanecieron iguales?

## 💡 Instrucciones adicionales
- Calcular los indicadores para ambos escenarios
- Comparar los resultados y determinar los porcentajes de cambio
- Analizar el impacto económico de la decisión
- Presentar conclusiones claras y fundamentadas

```python
# 🏭 **Análisis de Productividad y Utilidades en BBBC**
# -----------------------------------------------------
# 📌 Evaluamos el impacto del cambio en el programa de producción de BBBC
#    en términos de productividad multifactorial, productividad laboral y utilidades semanales.

# -----------------------------------------------------
# 📊 Definimos las variables según los datos del problema
# -----------------------------------------------------

# 🔹 Producción Normal
unidades_original = 2500  # Uniformes producidos por semana
costo_produccion_original = 120  # Costo por uniforme ($)
precio_venta = 200  # Precio de venta por uniforme ($)

# 👷 Distribución de empleados y horas trabajadas
empleados_turno1 = 70  # Empleados en el turno 1
empleados_turno2 = 30  # Empleados en el turno 2
horas_trabajadas_original = 40 * (empleados_turno1 + empleados_turno2)  # Horas totales trabajadas

# 💰 Costos Totales
costo_total_original = unidades_original * costo_produccion_original  # Costo total de producción ($)

# 🔺 Producción con Turnos Extendidos
unidades_sobrecarga = 4000  # Uniformes producidos por semana con turnos extendidos
costo_produccion_sobrecarga = 144  # Nuevo costo por uniforme ($)
horas_trabajadas_sobrecarga = 72 * (empleados_turno1 + empleados_turno2)  # Horas totales trabajadas en sobrecarga
costo_total_sobrecarga = unidades_sobrecarga * costo_produccion_sobrecarga  # Costo total de producción ($)

# -----------------------------------------------------
# 📊 Cálculo de Productividad Multifactorial (PMF)
# -----------------------------------------------------
pmf_original = unidades_original / costo_total_original  # Productividad original
pmf_sobrecarga = unidades_sobrecarga / costo_total_sobrecarga  # Productividad con turnos extendidos
cambio_pmf = ((pmf_sobrecarga - pmf_original) / pmf_original) * 100  # Cambio porcentual en PMF

# -----------------------------------------------------
# 👨‍🏭 Cálculo de Productividad Laboral (PL)
# -----------------------------------------------------
pl_original = unidades_original / horas_trabajadas_original  # Productividad laboral original
pl_sobrecarga = unidades_sobrecarga / horas_trabajadas_sobrecarga  # Productividad laboral con sobrecarga
cambio_pl = ((pl_sobrecarga - pl_original) / pl_original) * 100  # Cambio porcentual en PL

# -----------------------------------------------------
# 💰 Cálculo de Utilidades Semanales
# -----------------------------------------------------
utilidad_original = (precio_venta - costo_produccion_original) * unidades_original  # Utilidad semanal original ($)
utilidad_sobrecarga = (precio_venta - costo_produccion_sobrecarga) * unidades_sobrecarga  # Utilidad semanal con sobrecarga ($)
cambio_utilidad = ((utilidad_sobrecarga - utilidad_original) / utilidad_original) * 100  # Cambio porcentual en utilidad

# -----------------------------------------------------
# 📢 Mostramos los resultados
# -----------------------------------------------------
print("🏭 **Resultados del Análisis de Productividad y Utilidad**\n")

print("📊 1. Productividad Multifactorial (PMF):")
print(f"   - Productividad original: {pmf_original:.6f}")
print(f"   - Productividad con sobrecarga: {pmf_sobrecarga:.6f}")
print(f"   - Cambio en PMF: {cambio_pmf:.2f}%\n")

print("👨‍🏭 2. Productividad Laboral (PL):")
print(f"   - Productividad original: {pl_original:.6f}")
print(f"   - Productividad con sobrecarga: {pl_sobrecarga:.6f}")
print(f"   - Cambio en PL: {cambio_pl:.2f}%\n")

print("💰 3. Utilidad Semanal:")
print(f"   - Utilidad original: ${utilidad_original:.2f}")
print(f"   - Utilidad con sobrecarga: ${utilidad_sobrecarga:.2f}")
print(f"   - Cambio en utilidad: {cambio_utilidad:.2f}%\n")

# -----------------------------------------------------
# 🔍 Análisis de los Resultados
# -----------------------------------------------------
print("📌 **Conclusiones:**")
if cambio_pmf < 0:
    print(f"   - La productividad multifactorial disminuyó en {round(cambio_pmf, 2)}% 🚨.")
else:
    print(f"   - La productividad multifactorial aumentó en {round(cambio_pmf, 2)}% ✅.")

if cambio_pl < 0:
    print(f"   - La productividad laboral disminuyó en {round(cambio_pl, 2)}% 🚨.")
else:
    print(f"   - La productividad laboral aumentó en {round(cambio_pl, 2)}% ✅.")

if cambio_utilidad > 0:
    print(f"   - Las utilidades semanales aumentaron en {round(cambio_utilidad, 2)}% ✅.")
else:
    print(f"   - Las utilidades semanales disminuyeron en {round(cambio_utilidad, 2)}% 🚨.")

print("\n📊 **Interpretación:** A pesar de la reducción en la eficiencia, el aumento en la producción permitió incrementar las utilidades semanales.")
```
