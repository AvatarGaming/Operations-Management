# 📘 **Ejercicio 01**  

---

> ✨ **Objetivo:** Calcular y analizar la **productividad multifactorial** y la **productividad laboral** en un curso típico de **3 créditos** en la Universidad *Boehring*.  

---

## 🔍 **Datos del problema**  

🎓 La matrícula estudiantil en la Universidad *Boehring* es de **$150** por hora crédito por semestre.  
🏛️ El estado subsidia la universidad con **$100** por hora crédito por semestre.  
📚 El tamaño promedio de un grupo en un curso de **3 créditos** es de **50 estudiantes**.  

### 💰 **Costos asociados**  

- 🧑‍🏫 **Costos laborales**: **$4,000** por clase  
- 📖 **Costos de materiales**: **$20** por estudiante por clase  
- 🏢 **Costos generales** (_overhead_): **$25,000** por clase  

---

## ❓ **Problemas a resolver**  

### 🏆 **Parte A: Productividad multifactorial**  
🔹 ¿Cuál es la razón de productividad **multifactorial** para este proceso de curso?  

### 👩‍🏫 **Parte B: Productividad laboral**  
📆 Si los instructores trabajan un promedio de **14 horas por semana** durante **16 semanas** para cada clase de **3 créditos** con **50 estudiantes**, ¿cuál es la razón de productividad **laboral**?  

---

## 📊 **Tabla 1: Datos del problema**  

| 🔢 **Variable**                     | 📖 **Descripción**                                      | 💲 **Valor**   |
|--------------------------------------|---------------------------------------------------------|--------------:|
| 🎓 **Cuota escolar por hora crédito**  | Costo que pagan los estudiantes por hora crédito       | **$150**      |
| 🏛️ **Apoyo estatal por hora crédito**  | Cantidad que aporta el gobierno estatal por hora crédito | **$100**      |
| 👨‍🎓 **Tamaño promedio del grupo**       | Número de estudiantes por grupo                        | **50**        |
| 📚 **Número de créditos por curso**   | Número de créditos por curso                           | **3**         |

> 📝 **Nota:** Se debe calcular la productividad teniendo en cuenta todos los costos y beneficios asociados al curso.  

---


```r
# 📘 Ejercicio 01 - Productividad en un Curso Universitario

# 📊 Definición de variables
# Costos y beneficios del curso
cuota_escolar <- 150  # 💰 Costo por hora crédito pagado por estudiante (USD)
apoyo_estatal <- 100   # 🏛️ Subsidio estatal por hora crédito (USD)
tamano_grupo <- 50     # 👨‍🎓 Número de estudiantes por clase
creditos_curso <- 3    # 📚 Número de créditos por curso

# 🏷️ Costos asociados
costo_mano_obra <- 4000    # 🧑‍🏫 Costo laboral por clase (USD)
costo_materiales <- 20     # 📖 Costo de materiales por estudiante (USD)
costos_generales <- 25000  # 🏢 Costos generales por clase (USD)

# ⏳ Horas trabajadas por el instructor
total_semanas <- 16          # 📅 Duración del curso en semanas
horas_por_semana <- 14       # ⏱️ Horas de trabajo del instructor por semana
horas_trabajadas <- horas_por_semana * total_semanas  # ⏳ Total de horas trabajadas

# 💡 Cálculo del valor del producto (Outputs)
total_cuota_escolar <- cuota_escolar * creditos_curso * tamano_grupo  # Ingresos por cuota escolar
apoyo_estatal_total <- apoyo_estatal * creditos_curso * tamano_grupo  # Ingresos por subsidio estatal
valor_producto_total <- total_cuota_escolar + apoyo_estatal_total  # 💵 Ingresos totales del curso

# 💡 Cálculo del valor de los insumos (Inputs)
costo_materiales_total <- costo_materiales * tamano_grupo  # 📖 Costo total de materiales
valor_insumos_total <- costo_mano_obra + costo_materiales_total + costos_generales  # 💰 Costos totales

# 📈 Cálculo de la productividad multifactorial
# 🔹 Productividad multifactorial = Valor del Producto / Valor de los Insumos
productividad_multifactorial <- valor_producto_total / valor_insumos_total

# 📈 Cálculo de la productividad laboral
# 🔹 Productividad laboral = Valor del Producto / Horas trabajadas
productividad_laboral <- valor_producto_total / horas_trabajadas

# 📢 Impresión de resultados
cat("=== 📊 Resultados del Ejercicio ===\n\n")

cat("📌 Valor del Producto (Outputs):\n")
cat("🎓 Cuota Escolar Total: $", total_cuota_escolar, "\n")
cat("🏛️ Apoyo Estatal Total: $", apoyo_estatal_total, "\n")
cat("💵 Valor Total del Producto: $", valor_producto_total, "\n\n")

cat("📌 Valor de los Insumos (Inputs):\n")
cat("🧑‍🏫 Costo de Mano de Obra: $", costo_mano_obra, "\n")
cat("📖 Costo de Materiales Total: $", costo_materiales_total, "\n")
cat("🏢 Costos Generales: $", costos_generales, "\n")
cat("💰 Valor Total de los Insumos: $", valor_insumos_total, "\n\n")

cat("📌 Productividad Multifactorial:\n")
cat("⚖️ Productividad = Valor del Producto / Valor de los Insumos =", productividad_multifactorial, "\n\n")

cat("📌 Productividad Laboral:\n")
cat("⏳ Horas trabajadas por instructor: ", horas_trabajadas, " horas\n")
cat("💵 Ingresos por curso: $", valor_producto_total, "\n")
cat("📊 Productividad Laboral (USD por hora):", productividad_laboral, "\n")
```

# 👗 **Ejercicio 02: Productividad Laboral en la Industria Textil**  

---

> 📌 **Objetivo:** Calcular y analizar la **productividad laboral** en la producción de prendas de vestir en *Natalie Attire*.  

---

## 🏭 **Descripción del proceso de manufactura**  

👕 *Natalie Attire* es una empresa de moda que fabrica prendas de vestir.  
📅 Durante una semana en particular:  

- ⏳ **360 horas** de trabajo fueron empleadas.  
- 🧵 Se produjo un lote de **132 prendas**, de las cuales:  
  - ❌ **52 prendas** eran de **segunda** (con defectos).  
  - ✅ **80 prendas** fueron vendidas a precio completo.  

### 💰 **Precios de venta**  

- 🏷️ **Prendas de segunda**: **$90** cada una (vendidas en la tienda de fábrica).  
- 🛍️ **Prendas sin defectos**: **$200** cada una (vendidas en distribución minorista).  

---

## ❓ **Problema a resolver**  

### 📊 **Cálculo de productividad laboral**  
🔹 ¿Cuál es la **razón de productividad laboral** (_labor productivity ratio_) en este proceso de manufactura?  

---

## 📊 **Tabla 1: Datos del problema**  

| 🔢 **Variable**                        | 💲 **Valor**   |
|----------------------------------------|--------------:|
| 👕 **Número total de prendas**          | **132**       |
| ❌ **Número de prendas de segunda**     | **52**        |
| ✅ **Número de prendas restantes**      | **80**        |
| 🏷️ **Precio de venta por prenda de segunda** | **$90**        |
| 🛍️ **Precio de venta por prenda restante**  | **$200**       |
| ⏳ **Horas de trabajo empleadas**       | **360**       |

> 📝 **Nota:** Para calcular la productividad laboral, se debe considerar el valor total de producción generado en función de las horas de trabajo utilizadas.  

---


```r
# 👗 **Ejercicio 02: Productividad Laboral en la Industria Textil**

# 📌 Definir variables clave del problema
# Datos de entrada sobre la producción y costos en Natalie Attire
num_prendas_totales <- 132   # Total de prendas producidas 👕
num_prendas_segunda <- 52    # Prendas defectuosas (segunda calidad) ❌
num_prendas_buenas <- 80     # Prendas sin defectos ✅
precio_segunda <- 90         # Precio de venta por prenda de segunda 🏷️
precio_buenas <- 200         # Precio de venta por prenda sin defectos 🛍️
horas_trabajo <- 360         # Total de horas trabajadas ⏳

# 💰 Cálculo del valor total de la producción
# Se obtiene multiplicando la cantidad de cada tipo de prenda por su precio de venta
valor_prendas_segunda <- num_prendas_segunda * precio_segunda  # Valor total de prendas de segunda
valor_prendas_buenas <- num_prendas_buenas * precio_buenas      # Valor total de prendas sin defectos

# 🔢 Sumar ambos valores para obtener el ingreso total generado por la producción
valor_total_producto <- valor_prendas_segunda + valor_prendas_buenas

# 📊 Cálculo de la productividad laboral
# Fórmula: Productividad laboral = Valor total de la producción / Horas trabajadas
productividad_laboral <- valor_total_producto / horas_trabajo

# 📢 Mostrar los resultados en consola con formato claro
cat("---- 📊 Resultados ----\n")
cat("🧾 Valor total de prendas de segunda: $", valor_prendas_segunda, "\n")
cat("🧾 Valor total de prendas buenas: $", valor_prendas_buenas, "\n")
cat("💰 Valor total del producto: $", valor_total_producto, "\n")
cat("⏳ Horas de trabajo empleadas:", horas_trabajo, "horas\n")
cat("📈 Productividad laboral: $", round(productividad_laboral, 2), " por hora\n")
```

# 🎓 **Ejercicio 03: Análisis de Productividad Multifactorial y Laboral**  

---

> 📌 **Objetivo:** Evaluar cómo los cambios en la Universidad *Boehring* afectan la **productividad multifactorial** y la **productividad laboral**.  

---

## 📖 **Contexto: Cambios en la Universidad *Boehring***  

📢 El entrenador **Bijoum Toulouse** ha llevado al equipo de fútbol **Big Red Herrings** a varias temporadas decepcionantes.  
⚽ Solo una mejor estrategia de reclutamiento permitirá que el equipo vuelva a su forma ganadora.  

🔻 **Problemas actuales:**  
- 🎟️ Los fanáticos de la universidad no están dispuestos a pagar más de **$192** por un pase de temporada.  
- 📈 Para mejorar el reclutamiento, los **costos generales** (_overhead_) aumentarán de **$25,000** a **$30,000** por sección de clase.  
- 🏫 Para compensar estos costos, el tamaño promedio de las clases aumentará de **50 a 75 estudiantes**.  

📊 **Impacto en los costos:**  
- 💰 **Costos laborales** aumentarán a **$6,500** por curso de 3 créditos.  
- 📚 **Costos de materiales** serán de **$25** por estudiante por curso.  
- 🎓 **Matrícula** aumentará a **$200** por crédito, con un **apoyo estatal** de **$100** por crédito.  

---

## ❓ **Preguntas a Resolver**  

### 🔢 **Parte A: Productividad Multifactorial**  
📊 ¿Cuál es el **índice de productividad multifactorial** después de estos cambios?  
📉 En comparación con el resultado del **Problema Resuelto 1**, ¿la productividad **aumentó o disminuyó**?  

### 👨‍🏫 **Parte B: Productividad Laboral**  
⏳ Si los instructores trabajan **20 horas por semana** durante **16 semanas** para cada curso de **3 créditos** con **75 estudiantes**,  
¿Cuál será el **índice de productividad laboral**?  

---

## 📊 **Tabla 1: Datos del Problema**  

| 🔢 **Variable**                             | 💲 **Valor**  |
|---------------------------------------------|-------------:|
| 🏛️ **Costo general por sección (nuevo)**    | **$30,000**  |
| 🏛️ **Costo general por sección (anterior)** | **$25,000**  |
| 🎓 **Tamaño promedio de la clase**          | **75**       |
| 💼 **Costos laborales por curso**           | **$6,500**   |
| 📖 **Costos de materiales por estudiante**  | **$25**      |
| 🎟️ **Costo de matrícula por crédito**       | **$200**     |
| 🏛️ **Apoyo estatal por crédito**            | **$100**     |
| 📚 **Créditos por curso**                   | **3**        |
| ⏳ **Horas de trabajo del instructor**      | **320**      |

> 📝 **Nota:** Se analizará el impacto de estos cambios en la eficiencia del proceso educativo y en el uso de recursos.  

---



```r
# ========================================================
# 🎓 Análisis de Productividad Multifactorial y Laboral en R
# ========================================================

# 📌 Administración de Operaciones: Evaluamos cómo los cambios 
# en costos y horas de trabajo afectan la productividad.

# ----------------------------------
# 📊 Definición de Variables Iniciales
# ----------------------------------

# 🔢 Costos y parámetros originales
cuota_escolar_original <- 150     # 💰 Costo de matrícula por crédito
apoyo_estatal_original <- 100     # 🏛️ Apoyo estatal por crédito
tamano_grupo_original <- 50       # 🎓 Tamaño de la clase
creditos_curso <- 3               # 📚 Créditos por curso
costo_mano_obra_original <- 4000  # 💼 Costos laborales por curso
costo_generales_original <- 25000 # 🏛️ Costos generales por sección
costo_materiales_original <- 1000 # 📖 Costos de materiales
horas_trabajo_original <- 224     # ⏳ Horas de trabajo del instructor

# 🔢 Costos y parámetros ajustados tras los cambios
cuota_escolar_ajuste <- 200       # 💰 Nueva matrícula por crédito
apoyo_estatal_ajuste <- 100       # 🏛️ Apoyo estatal sin cambios
tamano_grupo_ajuste <- 75         # 🎓 Nuevo tamaño de clase
costo_mano_obra_ajuste <- 6500    # 💼 Nuevos costos laborales
costo_generales_ajuste <- 30000   # 🏛️ Nuevos costos generales
costo_materiales_ajuste <- 1875   # 📖 Nuevos costos de materiales
horas_trabajo_ajuste <- 320       # ⏳ Nuevas horas de trabajo

# ----------------------------------
# 📈 Cálculo del Valor del Producto (Ingresos Generados)
# ----------------------------------
valor_producto_original <- (cuota_escolar_original + apoyo_estatal_original) * creditos_curso * tamano_grupo_original
valor_producto_ajuste <- (cuota_escolar_ajuste + apoyo_estatal_ajuste) * creditos_curso * tamano_grupo_ajuste

# ----------------------------------
# 💵 Cálculo de los Insumos (Costos Totales)
# ----------------------------------
insumos_original <- costo_mano_obra_original + costo_generales_original + costo_materiales_original
insumos_ajuste <- costo_mano_obra_ajuste + costo_generales_ajuste + costo_materiales_ajuste

# ----------------------------------
# 📊 Cálculo del Índice de Productividad Multifactorial
# ----------------------------------
productividad_multifactorial_original <- valor_producto_original / insumos_original
productividad_multifactorial_ajuste <- valor_producto_ajuste / insumos_ajuste

# ----------------------------------
# ⏳ Cálculo de la Productividad Laboral (por Hora Trabajada)
# ----------------------------------
productividad_laboral_original <- valor_producto_original / horas_trabajo_original
productividad_laboral_ajuste <- valor_producto_ajuste / horas_trabajo_ajuste

# ----------------------------------
# 📉 Cálculo del Cambio Porcentual en la Productividad
# ----------------------------------
cambio_productividad_multifactorial <- ((productividad_multifactorial_ajuste - productividad_multifactorial_original) / productividad_multifactorial_original) * 100
cambio_productividad_laboral <- ((productividad_laboral_ajuste - productividad_laboral_original) / productividad_laboral_original) * 100

# ----------------------------------
# 📢 Resultados del Análisis
# ----------------------------------
cat("\n--- 📊 Resultados de Productividad ---\n")
cat("💰 Valor del Producto (Original):", valor_producto_original, "\n")
cat("💰 Valor del Producto (Ajustado):", valor_producto_ajuste, "\n")
cat("📊 Productividad Multifactorial Original:", round(productividad_multifactorial_original, 4), "\n")
cat("📊 Productividad Multifactorial Ajustada:", round(productividad_multifactorial_ajuste, 4), "\n")
cat("📉 Cambio en Productividad Multifactorial:", round(cambio_productividad_multifactorial, 2), "%\n")
cat("⏳ Productividad Laboral Original:", round(productividad_laboral_original, 4), "\n")
cat("⏳ Productividad Laboral Ajustada:", round(productividad_laboral_ajuste, 4), "\n")
cat("📉 Cambio en Productividad Laboral:", round(cambio_productividad_laboral, 2), "%\n")

# ========================================================
# 📌 CONCLUSIÓN:
# - 📈 Se observa un incremento en la productividad laboral y multifactorial.
# - 📊 El aumento de costos fue compensado por un mayor tamaño de grupo y eficiencia.
# - 🔍 Es crucial evaluar continuamente estos indicadores para optimizar recursos.
# ========================================================
```

# 🧺 **Ejercicio 04: Productividad Laboral en Suds and Duds Laundry**  

---

> 📌 **Objetivo:** Analizar la **productividad laboral** en el servicio de lavandería *Suds and Duds Laundry*.  

---

## 🏭 **Descripción del Proceso**  

👔 *Suds and Duds Laundry* es un servicio de lavandería que se especializa en el **lavado y planchado de camisas de vestir**.  
📅 Se ha registrado la producción semanal durante **cinco semanas**, considerando el **número de trabajadores, total de horas trabajadas y camisas procesadas**.  

---

## 📊 **Datos de Producción**  

| 📅 **Semana** | 👥 **Trabajadores**  | ⏳ **Total de horas** | 👔 **Camisas procesadas** |
|-------------|--------------------|---------------------:|-------------------------:|
| 1           | **Sud y Dud**       | **24**              | **68**                   |
| 2           | **Sud y Dud**       | **46**              | **130**                  |
| 3           | **Sud, Dud y Jud**  | **62**              | **152**                  |
| 4           | **Sud, Dud y Jud**  | **51**              | **125**                  |
| 5           | **Dud y Jud**       | **45**              | **131**                  |

---

## ❓ **Preguntas a Resolver**  

### 🔢 **Parte A: Cálculo de Productividad Laboral**  
📊 ¿Cuál es la **razón de productividad laboral** (_labor productivity ratio_) para cada semana?  

### 📈 **Parte B: Análisis del Patrón de Productividad**  
🔍 ¿Qué **patrón de productividad laboral** se puede identificar a partir de los datos?  
📉 ¿Cómo influye el número de trabajadores y las horas trabajadas en la eficiencia del servicio?  

> 📝 **Nota:** La productividad laboral se calculará como el número de camisas procesadas por hora trabajada.  

---



```r
# 🏭 **Cálculo de la Productividad Laboral - Suds and Duds Laundry**
# -----------------------------------------------------------
# 📌 La productividad laboral se calcula como:
#    Productividad = Producción (output) / Horas de trabajo (input)

# 📊 Definir los datos del problema
semanas <- c(1, 2, 3, 4, 5)  # 📅 Número de semanas
trabajadores <- c(2, 2, 3, 3, 2)  # 👥 Número de trabajadores por semana
horas_trabajo <- c(24, 46, 62, 51, 45)  # ⏳ Total de horas trabajadas por semana
produccion <- c(68, 130, 152, 125, 131)  # 👔 Camisas producidas por semana

# 📈 Calcular la productividad laboral (camisas por hora trabajada)
productividad <- produccion / horas_trabajo  

# 🖥️ Imprimir los resultados de productividad laboral
cat("\n📊 Productividad Laboral (camisas por hora):\n")
for (i in 1:length(semanas)) {
  cat(sprintf("📅 Semana %d: %.4f camisas/hora\n", semanas[i], productividad[i]))
}

# 🔄 Analizar los cambios en productividad entre semanas
cambio_productividad <- c(NA, diff(productividad) / productividad[-length(productividad)] * 100)

# 📉 Imprimir los cambios porcentuales de productividad
cat("\n📉 Cambio porcentual en productividad entre semanas:\n")
for (i in 2:length(semanas)) {
  cat(sprintf("📅 Semana %d vs Semana %d: %.2f%%\n", semanas[i], semanas[i-1], cambio_productividad[i]))
}

# 🧐 Análisis de productividad laboral
cat("\n🔍 Análisis de la Productividad Laboral:\n")
cat("- La productividad laboral disminuye en la Semana 3 cuando hay 3 trabajadores.\n")
cat("- Esto sugiere que el tercer trabajador no contribuye significativamente al rendimiento.\n")
cat("- Podría deberse a restricciones de espacio, equipos o flujo de trabajo ineficiente.\n")
```

# 🔧 **Ejercicio 05: Optimización de Costos en White Tiger Electronics**  

---

> 📌 **Objetivo:** Determinar los **porcentajes de reducción** necesarios en los costos de **materiales, mano de obra y costos generales**  
para lograr una **mejora del 10% en la productividad multifactorial** en *White Tiger Electronics*.  

---

## 🏭 **Descripción del Proceso**  

📀 *White Tiger Electronics* fabrica **reproductores de CD** utilizando un **proceso de ensamblaje automatizado**.  
💰 El **costo estándar** de producción por unidad es de **$150**, distribuido de la siguiente manera:  

---

## 📊 **Distribución de Costos**  

| 💼 **Concepto**      | 💲 **Costo ($)** |
|---------------------|---------------:|
| 👷 **Mano de obra**  | **30**         |
| 🛠️ **Materiales**   | **70**         |
| 🏭 **Costos generales** | **50**         |
| 💰 **Total**         | **150**        |

🏷️ **Precio de venta por unidad:** **$300**  

---

## ❓ **Preguntas a Resolver**  

### 🔢 **Parte A: Reducción de Costos en Materiales**  
📉 ¿En qué **porcentaje deben reducirse los costos de materiales** para lograr una **mejora del 10% en la productividad multifactorial**?  

### 👷 **Parte B: Reducción de Costos en Mano de Obra**  
📉 ¿En qué **porcentaje deben reducirse los costos de mano de obra** para alcanzar la misma mejora en productividad?  

### 🏭 **Parte C: Reducción de Costos Generales**  
📉 ¿En qué **porcentaje deben reducirse los costos generales** para alcanzar el objetivo de mejora?  

> 📝 **Nota:** Se analizará la **relación entre los costos totales y la productividad** para determinar el impacto de cada tipo de reducción.  

---


```r
# 📌 **Ejercicio 05: Optimización de Costos en White Tiger Electronics**
# =========================================================
# 🔧 Cálculo de reducción de costos para mejorar productividad multifactorial
# 📊 Administración de operaciones - Análisis de eficiencia en costos
# =========================================================

# 📌 Definición de variables clave
precio_venta <- 300       # 💰 Precio de venta por unidad ($)
costo_total <- 150        # 🏷️ Costo total de producción por unidad ($)
costo_mano_obra <- 30     # 👷 Costo de mano de obra por unidad ($)
costo_materiales <- 70    # 🛠️ Costo de materiales por unidad ($)
costo_overhead <- 50      # 🏭 Costos generales (Overhead) por unidad ($)
mejora_productividad <- 1.1  # 📈 Mejora esperada del 10% en productividad multifactorial

# 📊 Cálculo de la nueva entrada de recursos requerida
nuevo_input <- costo_total / mejora_productividad

# 📉 Cálculo de la reducción total necesaria en costos
reduccion_necesaria <- costo_total - nuevo_input

# 🔢 Cálculo de reducción en Materiales
reduccion_materiales <- (reduccion_necesaria / costo_materiales) * 100
nuevo_costo_materiales <- costo_materiales - reduccion_necesaria

# 🔢 Cálculo de reducción en Mano de Obra
reduccion_mano_obra <- (reduccion_necesaria / costo_mano_obra) * 100
nuevo_costo_mano_obra <- costo_mano_obra - reduccion_necesaria

# 🔢 Cálculo de reducción en Costos Generales (Overhead)
reduccion_overhead <- (reduccion_necesaria / costo_overhead) * 100
nuevo_costo_overhead <- costo_overhead - reduccion_necesaria

# 📋 Impresión de resultados
cat("=======================================\n")
cat(" White Tiger Electronics - Resultados \n")
cat("=======================================\n\n")

cat("🏷️ Costo total original de producción por unidad:", costo_total, "\n")
cat("📊 Nuevo costo total necesario para lograr mejora del 10%:", round(nuevo_input, 4), "\n")
cat("📉 Reducción total requerida en costos:", round(reduccion_necesaria, 4), "\n\n")

cat("🛠️ Reducción necesaria en Materiales:\n")
cat(" - Nuevo costo de materiales:", round(nuevo_costo_materiales, 4), "\n")
cat(" - Porcentaje de reducción:", round(reduccion_materiales, 2), "%\n\n")

cat("👷 Reducción necesaria en Mano de Obra:\n")
cat(" - Nuevo costo de mano de obra:", round(nuevo_costo_mano_obra, 4), "\n")
cat(" - Porcentaje de reducción:", round(reduccion_mano_obra, 2), "%\n\n")

cat("🏭 Reducción necesaria en Costos Generales (Overhead):\n")
cat(" - Nuevo costo de overhead:", round(nuevo_costo_overhead, 4), "\n")
cat(" - Porcentaje de reducción:", round(reduccion_overhead, 2), "%\n")
```

# 🏭 **Ejercicio 06: Análisis de Mejoras en el Proceso de Symtecks**  

---

> 📌 **Objetivo:** Evaluar el impacto de las mejoras recientes en un proceso de **Symtecks**,  
utilizando **productividad multifactorial** y **productividad laboral** como indicadores de desempeño.  

---

## 🏢 **Descripción del Proceso**  

⚙️ En **Symtecks**, un proceso específico genera un **output** con un valor de **$100 por unidad producida**.  
👷‍♂️ El **costo de la mano de obra** es de **$50 por hora**, incluyendo beneficios.  

📊 **El departamento de contabilidad** ha proporcionado información sobre la producción y costos en las últimas **cuatro semanas**.  

---

## 📊 **Datos del Proceso**  

| 📅 **Concepto**       | 🔹 **Week 1** | 🔹 **Week 2** | 🔹 **Week 3** | 🔹 **Week 4** |
|----------------------|-------------:|-------------:|-------------:|-------------:|
| 📦 **Unidades Producidas** | **1,124** | **1,310** | **1,092** | **981** |
| 👷‍♂️ **Mano de Obra ($)** | **12,735** | **14,842** | **10,603** | **9,526** |
| 🛠️ **Materiales ($)** | **21,041** | **24,523** | **20,442** | **18,364** |
| 🏭 **Overhead ($)** | **8,992** | **10,480** | **8,736** | **7,848** |

---

## ❓ **Preguntas a Resolver**  

### 🔢 **Parte A: Productividad Multifactorial**  
📊 ¿Las **mejoras recientes** en el proceso han tenido algún impacto en la productividad multifactorial?  
📈 ¿En qué semana se observa el **mayor cambio** en la productividad multifactorial?  

### 👨‍🏭 **Parte B: Productividad Laboral**  
📉 ¿La **productividad laboral** ha cambiado a lo largo de las semanas?  
🧮 Utilizar la razón de **productividad laboral** para justificar la respuesta.  

> 📝 **Nota:** Se evaluará la **eficiencia del uso de recursos** y el impacto de las mejoras en la producción.  

---


```r
# 🏭 **Cálculo de Productividad Multifactorial y Laboral**
# -------------------------------------------------------
# 📌 Evaluamos la eficiencia del proceso en Symtecks mediante dos métricas clave:
#    1️⃣ Productividad Multifactorial (PMF): Relación entre el valor de producción y el costo total de insumos.
#    2️⃣ Productividad Laboral (PL): Relación entre unidades producidas y horas trabajadas.

# 📊 **Datos Iniciales**
valor_output <- 100  # Precio por unidad producida ($)
costo_laboral <- 50   # Costo por hora de trabajo ($)

# 🔹 **Datos Semanales**
unidades_producidas <- c(1124, 1310, 1092, 981)
costo_laboral_total <- c(12735, 14842, 10603, 9526)
horas_laborales <- c(254.7, 296.84, 212.06, 190.52)
materiales <- c(21041, 24523, 20442, 18364)
overhead <- c(8992, 10480, 8736, 7848)

# 🔍 **Cálculos**
# -------------------------------------------------------
# 📌 **Paso 1: Calcular el valor total de producción**
valor_total <- unidades_producidas * valor_output

# 📌 **Paso 2: Calcular el insumo total** (suma de costos laborales, materiales y overhead)
insumo_total <- costo_laboral_total + materiales + overhead

# 📌 **Paso 3: Calcular la Productividad Multifactorial (PMF)**
productividad_multifactorial <- valor_total / insumo_total

# 📌 **Paso 4: Calcular la Productividad Laboral (PL)**
productividad_laboral <- unidades_producidas / horas_laborales

# 📊 **Resultados**
# -------------------------------------------------------
cat("\n--- 📈 Resultados de Productividad ---\n")
cat("Semana | Valor Total | Insumo Total | PMF    | Horas  | PL    \n")
cat("----------------------------------------------------------\n")
for (i in 1:4) {
  cat(sprintf("  %d    | %10.2f | %12.2f | %.4f | %.2f | %.4f \n",
              i, valor_total[i], insumo_total[i], productividad_multifactorial[i], 
              horas_laborales[i], productividad_laboral[i]))
}

# 🔍 **Análisis de Mejora**
# -------------------------------------------------------
mejora_pmf <- ((productividad_multifactorial[4] - productividad_multifactorial[1]) / productividad_multifactorial[1]) * 100
mejora_pl <- ((productividad_laboral[4] - productividad_laboral[1]) / productividad_laboral[1]) * 100

cat("\n--- 📊 Análisis de Cambio en Productividad ---\n")
cat(sprintf("📌 Cambio en Productividad Multifactorial: %.2f%%\n", mejora_pmf))
cat(sprintf("📌 Cambio en Productividad Laboral: %.2f%%\n", mejora_pl))

# 🏁 **Conclusión**
# -------------------------------------------------------
if (mejora_pmf > 0) {
  cat("\n✅ La productividad multifactorial ha mejorado a lo largo de las semanas.\n")
} else {
  cat("\n❌ No hubo mejora significativa en la productividad multifactorial.\n")
}

if (mejora_pl > 0) {
  cat("✅ La productividad laboral también ha aumentado, indicando una mayor eficiencia del trabajo.\n")
} else {
  cat("❌ No hubo mejora significativa en la productividad laboral.\n")
}
```

# 🏭 **Ejercicio 07: Análisis de Productividad y Utilidades en BBBC**  

---

> 📌 **Objetivo:** Evaluar el impacto de un cambio en el **programa de producción** de **Big Black Bird Company (BBBC)**  
en términos de **productividad multifactorial, productividad laboral y utilidades semanales**.  

---

## 🏢 **Descripción del Problema**  

🦅 **Big Black Bird Company (BBBC)** recibió un **pedido urgente** de uniformes militares con un **revestimiento especial de plástico**.  
📈 Debido a la urgencia, se modificó el programa de producción, aumentando las horas de trabajo semanales.  

---

## 🔹 **Producción Normal**  

- **📆 Turnos:** **2 turnos** de **40 horas** cada uno (por semana).  
- **👕 Producción:** **2,500 uniformes/semana**.  
- **💰 Costo estándar por uniforme:** **$120**.  
- **👷 Distribución de empleados:**  
  - 🏭 **70 empleados** en el **primer turno**.  
  - 🏭 **30 empleados** en el **segundo turno**.  
- **📌 Precio de contrato:** **$200 por uniforme**.  

---

## 🔺 **Producción con Turnos Extendidos**  

- 🕐 **Cada turno trabaja 72 horas por semana**.  
- 📈 **Producción aumenta a 4,000 uniformes/semana**.  
- 💲 **Costo por uniforme aumenta a $144**.  

---

## ❓ **Preguntas a Resolver**  

### 🔢 **Parte A: Productividad Multifactorial**  
📊 ¿La **productividad multifactorial** aumentó, disminuyó o se mantuvo igual?  
📈 Si cambió, ¿en qué **porcentaje**?  

### 👨‍🏭 **Parte B: Productividad Laboral**  
📉 ¿La **productividad laboral** cambió con la extensión de turnos?  
🧮 Si hubo un cambio, ¿cuál fue el **porcentaje de variación**?  

### 💰 **Parte C: Utilidades Semanales**  
📈 ¿Las **utilidades semanales** aumentaron, disminuyeron o se mantuvieron iguales tras el cambio de producción?  

> 📝 **Nota:** Se evaluará la eficiencia del **uso de recursos, costos y beneficios** asociados con el nuevo programa.  

---



```r
# 🏭 **Análisis de Productividad y Utilidades en BBBC**
# -----------------------------------------------------
# 📌 Evaluamos el impacto del cambio en el programa de producción de BBBC
#    en términos de productividad multifactorial, productividad laboral y utilidades semanales.

# -----------------------------------------------------
# 📊 Definimos las variables según los datos del problema
# -----------------------------------------------------

# 🔹 Producción Normal
unidades_original <- 2500  # Uniformes producidos por semana
costo_produccion_original <- 120  # Costo por uniforme ($)
precio_venta <- 200  # Precio de venta por uniforme ($)

# 👷 Distribución de empleados y horas trabajadas
empleados_turno1 <- 70  # Empleados en el turno 1
empleados_turno2 <- 30  # Empleados en el turno 2
horas_trabajadas_original <- 40 * (empleados_turno1 + empleados_turno2)  # Horas totales trabajadas

# 💰 Costos Totales
costo_total_original <- unidades_original * costo_produccion_original  # Costo total de producción ($)

# 🔺 Producción con Turnos Extendidos
unidades_sobrecarga <- 4000  # Uniformes producidos por semana con turnos extendidos
costo_produccion_sobrecarga <- 144  # Nuevo costo por uniforme ($)
horas_trabajadas_sobrecarga <- 72 * (empleados_turno1 + empleados_turno2)  # Horas totales trabajadas en sobrecarga
costo_total_sobrecarga <- unidades_sobrecarga * costo_produccion_sobrecarga  # Costo total de producción ($)

# -----------------------------------------------------
# 📊 Cálculo de Productividad Multifactorial (PMF)
# -----------------------------------------------------
pmf_original <- unidades_original / costo_total_original  # Productividad original
pmf_sobrecarga <- unidades_sobrecarga / costo_total_sobrecarga  # Productividad con turnos extendidos
cambio_pmf <- ((pmf_sobrecarga - pmf_original) / pmf_original) * 100  # Cambio porcentual en PMF

# -----------------------------------------------------
# 👨‍🏭 Cálculo de Productividad Laboral (PL)
# -----------------------------------------------------
pl_original <- unidades_original / horas_trabajadas_original  # Productividad laboral original
pl_sobrecarga <- unidades_sobrecarga / horas_trabajadas_sobrecarga  # Productividad laboral con sobrecarga
cambio_pl <- ((pl_sobrecarga - pl_original) / pl_original) * 100  # Cambio porcentual en PL

# -----------------------------------------------------
# 💰 Cálculo de Utilidades Semanales
# -----------------------------------------------------
utilidad_original <- (precio_venta - costo_produccion_original) * unidades_original  # Utilidad semanal original ($)
utilidad_sobrecarga <- (precio_venta - costo_produccion_sobrecarga) * unidades_sobrecarga  # Utilidad semanal con sobrecarga ($)
cambio_utilidad <- ((utilidad_sobrecarga - utilidad_original) / utilidad_original) * 100  # Cambio porcentual en utilidad

# -----------------------------------------------------
# 📢 Mostramos los resultados
# -----------------------------------------------------
cat("🏭 **Resultados del Análisis de Productividad y Utilidad**\n\n")

cat("📊 1. Productividad Multifactorial (PMF):\n")
cat(sprintf("   - Productividad original: %.6f\n", pmf_original))
cat(sprintf("   - Productividad con sobrecarga: %.6f\n", pmf_sobrecarga))
cat(sprintf("   - Cambio en PMF: %.2f%%\n\n", cambio_pmf))

cat("👨‍🏭 2. Productividad Laboral (PL):\n")
cat(sprintf("   - Productividad original: %.6f\n", pl_original))
cat(sprintf("   - Productividad con sobrecarga: %.6f\n", pl_sobrecarga))
cat(sprintf("   - Cambio en PL: %.2f%%\n\n", cambio_pl))

cat("💰 3. Utilidad Semanal:\n")
cat(sprintf("   - Utilidad original: $%.2f\n", utilidad_original))
cat(sprintf("   - Utilidad con sobrecarga: $%.2f\n", utilidad_sobrecarga))
cat(sprintf("   - Cambio en utilidad: %.2f%%\n\n", cambio_utilidad))

# -----------------------------------------------------
# 🔍 Análisis de los Resultados
# -----------------------------------------------------
cat("📌 **Conclusiones:**\n")
if (cambio_pmf < 0) {
  cat("   - La productividad multifactorial disminuyó en", round(cambio_pmf, 2), "% 🚨.\n")
} else {
  cat("   - La productividad multifactorial aumentó en", round(cambio_pmf, 2), "% ✅.\n")
}

if (cambio_pl < 0) {
  cat("   - La productividad laboral disminuyó en", round(cambio_pl, 2), "% 🚨.\n")
} else {
  cat("   - La productividad laboral aumentó en", round(cambio_pl, 2), "% ✅.\n")
}

if (cambio_utilidad > 0) {
  cat("   - Las utilidades semanales aumentaron en", round(cambio_utilidad, 2), "% ✅.\n")
} else {
  cat("   - Las utilidades semanales disminuyeron en", round(cambio_utilidad, 2), "% 🚨.\n")
}

cat("\n📊 **Interpretación:** A pesar de la reducción en la eficiencia, el aumento en la producción permitió incrementar las utilidades semanales.\n")
```
